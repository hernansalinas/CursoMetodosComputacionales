## Librería en Python

Una **librería en Python** es un conjunto de módulos o paquetes que contienen funciones, clases y métodos predefinidos que facilitan la realización de tareas específicas sin la necesidad de reescribir código desde cero. Estas librerías permiten a los desarrolladores aprovechar soluciones optimizadas y probadas, lo que mejora la productividad y la calidad del código.

### Importancia de las Librerías en Python

- **Facilitan el desarrollo**: Al reutilizar funciones y módulos predefinidos, los desarrolladores pueden centrarse en la lógica específica de sus proyectos sin preocuparse por la implementación de bajo nivel.
- **Ahorro de tiempo**: Evitan la necesidad de escribir código repetitivo o complejo desde cero.
- **Soluciones robustas**: Las librerías son mantenidas por comunidades activas, lo que garantiza que los errores sean rápidamente corregidos y que nuevas características se añadan constantemente.
- **Compatibilidad**: Muchas librerías están optimizadas para interactuar con otras, lo que permite una integración sencilla en proyectos más grandes.

### Funciones de una Librería en Python

1. **Funciones predefinidas**: Proporcionan bloques de código reutilizables para realizar tareas comunes como operaciones matemáticas, manipulación de cadenas, etc.
2. **Modularidad**: Permiten organizar el código en módulos más pequeños y manejables, lo que facilita su mantenimiento y reutilización.
3. **Abstracción**: Esconden la complejidad detrás de interfaces simples, lo que permite a los usuarios utilizar funciones complejas sin tener que entender cada detalle técnico.

## NumPy

**NumPy** (Numerical Python) es una librería fundamental para la computación científica en Python. Proporciona soporte para arreglos y matrices multidimensionales, junto con una colección de funciones matemáticas de alto rendimiento para operar sobre estos datos.

### Importancia de NumPy

- **Eficiencia**: NumPy está diseñado para trabajar con grandes volúmenes de datos de manera eficiente. Su implementación en C permite que las operaciones sean mucho más rápidas en comparación con las listas estándar de Python.
- **Facidad de uso**: Ofrece una sintaxis sencilla y clara que facilita el manejo y la manipulación de datos.
- **Integración**: Es compatible con otras librerías populares como Pandas, Matplotlib y SciPy, lo que la convierte en una herramienta versátil para el análisis de datos y la visualización.
- **Funcionalidades avanzadas**: Permite realizar operaciones matemáticas y lógicas complejas sobre arreglos de manera sencilla, lo que es fundamental para el análisis de datos y el aprendizaje automático.

### Funciones Básicas de NumPy

1. **Creación de arreglos**
   - **`numpy.array()`**: Crea un arreglo a partir de una lista o tupla.
     ```python
     import numpy as np
     arr = np.array([1, 2, 3, 4, 5])
     ```

2. **Operaciones matemáticas**
   - **`numpy.mean()`**: Calcula el promedio de los elementos de un arreglo.
     ```python
     promedio = np.mean(arr)
     print(promedio)  # Salida: 3.0
     ```

3. **Indexación y segmentación**
   - Permite acceder a elementos específicos o segmentos de un arreglo.
     ```python
     print(arr[0])       # Primer elemento: 1
     print(arr[1:4])     # Elementos del índice 1 al 3: [2 3 4]
     ```

4. **Funciones estadísticas**
   - **`numpy.std()`**: Calcula la desviación estándar de los elementos de un arreglo.
     ```python
     desviacion_estandar = np.std(arr)
     print(desviacion_estandar)  # Salida: 1.4142135623730951
     ```

5. **Operaciones lógicas**
   - Permite aplicar condiciones a los elementos de un arreglo.
     ```python
     arr_mayores_dos = arr[arr > 2]
     print(arr_mayores_dos)  # Salida: [3 4 5]
     ```




````python
arr = np.array([1, 2, 3, 4, 5])
promedio = np.mean(arr)
std = np.std(arr)
arr_mayores_dos = arr[arr > 2]


print(f"Promedio: {promedio}")
print(f"Desviación estándar: {std:.2f}")
print(f"Elementos mayores a 2: {arr_mayores_dos}")
````

## Matplotlib

**Matplotlib** es una librería de visualización de datos en Python que permite crear gráficos en 2D de alta calidad de manera sencilla. Es ampliamente utilizada en análisis de datos, ciencias y cualquier campo que requiera la representación visual de información.

### Importancia de Matplotlib

- **Visualización efectiva**: Permite crear gráficos atractivos y claros que facilitan la comprensión de los datos.
- **Flexibilidad**: Ofrece una amplia variedad de tipos de gráficos, desde gráficos de líneas y dispersión hasta histogramas y gráficos de barras.
- **Personalización**: Los gráficos pueden ser personalizados en gran medida, lo que permite a los usuarios ajustar estilos, colores y anotaciones según sus necesidades.
- **Integración**: Se integra fácilmente con otras librerías populares como NumPy y Pandas, lo que facilita la visualización de datos complejos.

### Funciones Básicas de Matplotlib

1. **Crear un gráfico de líneas**
   - La función más básica para trazar datos.
     ```python
     import matplotlib.pyplot as plt

     x = [1, 2, 3, 4, 5]
     y = [10, 20, 25, 30, 35]
     plt.plot(x, y)
     plt.title("Gráfico de líneas")
     plt.xlabel("Eje X")
     plt.ylabel("Eje Y")
     plt.show()
     ```

2. **Gráfico de barras**
   - Útil para comparar diferentes categorías.
     ```python
     categorias = ['A', 'B', 'C', 'D']
     valores = [3, 7, 5, 8]
     plt.bar(categorias, valores)
     plt.title("Gráfico de barras")
     plt.xlabel("Categorías")
     plt.ylabel("Valores")
     plt.show()
     ```

3. **Histograma**
   - Muestra la distribución de un conjunto de datos.
     ```python
     datos = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5, 5]
     plt.hist(datos, bins=5, alpha=0.7, color='blue')
     plt.title("Histograma")
     plt.xlabel("Valores")
     plt.ylabel("Frecuencia")
     plt.show()
     ```

4. **Gráfico de dispersión**
   - Muestra la relación entre dos variables.
     ```python
     x = [5, 7, 8, 6, 9]
     y = [7, 10, 6, 9, 12]
     plt.scatter(x, y)
     plt.title("Gráfico de dispersión")
     plt.xlabel("Eje X")
     plt.ylabel("Eje Y")
     plt.show()
     ```

5. **Personalización de gráficos**
   - Cambiar el estilo de los gráficos.
     ```python
     plt.plot(x, y, marker='o', linestyle='--', color='green')
     plt.title("Gráfico personalizado")
     plt.xlabel("Eje X")
     plt.ylabel("Eje Y")
     plt.grid()
     plt.show()
     ```



````python
# Gráfico lineas
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
plt.plot(x, y, 'r--')
plt.title("Gráfico de dispersión")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
````

````python
# Gráfico de barras
categorias = ['A', 'B', 'C', 'D']
valores = [3, 7, 5, 8]
plt.bar(categorias, valores)
plt.title("Gráfico de barras")
plt.xlabel("Categorías")
plt.ylabel("Valores")
plt.show()
````

# Creación de Múltiples Gráficas en Matplotlib

Matplotlib es una poderosa librería de visualización de datos en Python que permite crear múltiples gráficas en una sola figura. Esto es especialmente útil para comparar diferentes conjuntos de datos o para visualizar varias representaciones de la misma información. A continuación, se describen los pasos básicos para crear múltiples gráficas.

## Uso de `plt.subplots()`

El método `plt.subplots()` es la forma más común de crear múltiples gráficos. Este método permite definir una cuadrícula de subgráficas en una figura. La función devuelve una figura y un array de ejes que se pueden utilizar para trazar diferentes gráficos.

### Sintaxis Básica

```python
fig, axs = plt.subplots(nrows, ncols)


````python
import matplotlib.pyplot as plt

# Crear una figura con 1 fila y 2 columnas
fig, axs = plt.subplots(1, 2)

# Gráfico 1: Gráfico de líneas
axs[0].plot([1, 2, 3], [1, 4, 9], 'ro-')
axs[0].set_title("Gráfico de Líneas")
axs[0].set_xlabel("Eje X")
axs[0].set_ylabel("Eje Y")

# Gráfico 2: Gráfico de barras
axs[1].bar(['A', 'B', 'C'], [3, 7, 5])
axs[1].set_title("Gráfico de Barras")
axs[1].set_xlabel("Categorías")
axs[1].set_ylabel("Valores")

# Ajustar el diseño
plt.tight_layout()
plt.show()
````
