$e^x \sin(x)=x$

```python
f = lambda x:np.exp(x)*np.sin(x)-x
x = np.linspace(3.5,9)
```

```python
plt.plot(x, f(x))
plt.grid()
```

```python
f = lambda x: (x-0.9)**2-1
x=np.linspace(-2, 4, 1000)
plt.plot(x, f(x))
plt.grid()
```

```python
xa = 1
xb = 6
epsilon=1e-5
Nmax = 100
def bisection(f, xa, xb, epsilon, Nmax):
  niter = 0
  while (f(xa)*f(xb)<0 | niter<Nmax):
    #print("existe un cero")
    niter = niter+1
    xc = (xa+xb)*0.5
    if (f(xc)<0):
      xa = xc
    elif(f(xc) >0):
      xb = xc
    elif( abs(f(xc))<epsilon ):
      #print(f"El cero es {xc}")
      break
    else :  # f(xc=0)
      print("Error")
  return xc
bisection(f, xa, xb, epsilon, Nmax)
```
