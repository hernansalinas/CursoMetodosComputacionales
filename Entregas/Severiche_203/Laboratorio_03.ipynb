## 1) Calcular de 2 formas distintas la siguiente expresi贸n:

\begin{equation}
e^{-x}
\end{equation}

```python
P1_x = 3/10
P1_n = 170
```

### Forma 1

\begin{equation}
e^{-x}=\sum_{n=0}^{N}  (-1)^n \frac{x^{n}}{n!}
\end{equation}

```python
def ExpNeg(x,n):
    result = 0.0
    for i in range(n):
        result = result + ((-1)**i)*(x**i/math.factorial(i))
    return result
```

```python
P1_exp_F1 = ExpNeg(P1_x,P1_n)
print(P1_exp_F1)
```

### Forma 2

Calculamos el valor:
\begin{equation}
e^{x}=\sum_{n=0}^{N} \frac{x^{n}}{n!}
\end{equation}

Posteriormente calculamos el valor objetivo:
\begin{equation}
e^{-x} = \frac{1}{e^{x}}
\end{equation}

```python
def Exp(x,n):
    result = 0.0
    for i in range(n):
        result = result + (x**i/math.factorial(i))
    return 1/result
```

```python
P1_exp_F2 = Exp(P1_x,P1_n)
print(P1_exp_F2)
```

### Comparaci贸n

\begin{equation}
abs(e^{-x} - e^{x})
\end{equation}

```python
print("La diferencia entre los calculos es:")
print(abs(P1_exp_F1-P1_exp_F2))
```

## 6) Calcular para un valor de x menor o igual a 1, donde la precisi贸n no pase del orden de 10^-7

\begin{equation}
\left| \frac{(-x)^{N+1}}{(N+1)!} \right | \le \left| 10^{-7} \sum_{N=0}^{N} \frac{(-x)^{n}}{n!} \right|
\end{equation}

```python
def Func_izq(x,N):
    return abs(((-x)**(N+1))/math.factorial(N+1))
```

```python
def Func_der(x,N):
    result = 0
    for i in range(0,N):
        result = ((-x)**i)/math.factorial(i)
    return abs(result)
```

```python
P6_valx = np.linspace(0,1,1000)
P6_valn = np.linspace(1,30,30)
```

```python
P6_dif = np.zeros(len(P6_valx))
for i in range(1,len(P6_valx)):
    for j in range(0,len(P6_valn)):
        lado_izq = Func_izq(P6_valx[i],int(P6_valn[j]))
        lado_der = Func_der(P6_valx[i],int(P6_valn[j]))
        if lado_izq <= lado_der*(10**(-7)):
            print("El primer valor de x que cumple la condici贸n es:",P6_valx[i])
            break
```
