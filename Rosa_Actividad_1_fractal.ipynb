Primero seguí los pasos, directamente en python para saber mas o menos que hacer

````python
import random
import matplotlib.pyplot as plt

# Inicialización
x, y = 0, 0
x_vals, y_vals = [], []

# Generar puntos
n_points = 50000

for _ in range(n_points):
    r = random.random()
    if r <= 0.01:
        #  regla1
        x_nuevo = 0
        y_nuevo = 0.16 * y
    elif r <= 0.86:
        #  regla2
        x_nuevo = 0.85 * x + 0.04 * y
        y_nuevo = -0.04 * x + 0.85 * y + 1.6
    elif r <= 0.93:
        #  regla3
        x_nuevo = 0.2 * x - 0.26 * y
        y_nuevo = 0.23 * x + 0.22 * y + 1.6
    else:
        #  regla4
        x_nuevo = -0.15 * x + 0.28 * y
        y_nuevo = 0.26 * x + 0.24 * y + 0.44

    x_vals.append(x_nuevo)
    y_vals.append(y_nuevo)

    x, y = x_nuevo, y_nuevo

plt.figure(figsize=(6, 10))
plt.scatter(x_vals, y_vals, s=0.1, color='green')
plt.title("Helecho de Barnsley")
plt.axis('off')
plt.show()
````

Codigo en Bash:

````python
#!/bin/bash

x=0
y=0
output_file="helecho_bash.txt"
n_points=50000


echo "" > $output_file #archivo de salida

for (( i=0; i<$n_points; i++ ))
do
    r=$(echo "scale=6; $RANDOM/32767" | bc)

    if (( $(echo "$r <= 0.01" | bc -l) ))
    then
        x_nuevo=0
        y_nuevo=$(echo "scale=6; 0.16 * $y" | bc)
    elif (( $(echo "$r <= 0.86" | bc -l) ))
    then
        x_nuevo=$(echo "scale=6; 0.85 * $x + 0.04 * $y" | bc)
        y_nuevo=$(echo "scale=6; -0.04 * $x + 0.85 * $y + 1.6" | bc)
    elif (( $(echo "$r <= 0.93" | bc -l) ))
    then
        x_nuevo=$(echo "scale=6; 0.2 * $x - 0.26 * $y" | bc)
        y_nuevo=$(echo "scale=6; 0.23 * $x + 0.22 * $y + 1.6" | bc)
    else
        x_nuevo=$(echo "scale=6; -0.15 * $x + 0.28 * $y" | bc)
        y_nuevo=$(echo "scale=6; 0.26 * $x + 0.24 * $y + 0.44" | bc)
    fi

    # Guardar valores
    echo "$x_nuevo $y_nuevo" >> $output_file

    # Actualizar
    x=$x_nuevo
    y=$y_nuevo
done
````

Codigo en Julia:

````python
x, y = 0.0, 0.0
n_points = 50000
output_file_julia = "helecho_julia.txt"

# Crear puntos
with open(output_file_julia, 'w') as f:
    for _ in range(n_points):
        r = random.random()
        if r <= 0.01:
            x_nuevo, y_nuevo = 0.0, 0.16 * y
        elif r <= 0.86:
            x_nuevo = 0.85 * x + 0.04 * y
            y_nuevo = -0.04 * x + 0.85 * y + 1.6
        elif r <= 0.93:
            x_nuevo = 0.2 * x - 0.26 * y
            y_nuevo = 0.23 * x + 0.22 * y + 1.6
        else:
            x_nuevo = -0.15 * x + 0.28 * y
            y_nuevo = 0.26 * x + 0.24 * y + 0.44

        # Guardar valors
        f.write(f"{x_nuevo} {y_nuevo}\n")
        x, y = x_nuevo, y_nuevo
````

Impresionantemente, es compatible con Colab, asi que usaremos el de Julia, ahora a leer los archivos con codigo python.

````python
import matplotlib.pyplot as plt

# lee archivo y carga puntos
def leer_puntos(archivo):
    x_vals, y_vals = [], []
    with open(archivo, 'r') as f:
        for linea in f:
            x, y = map(float, linea.split())
            x_vals.append(x)
            y_vals.append(y)
    return x_vals, y_vals

# Leer puntos generados en Julia
x_vals_bash, y_vals_bash = leer_puntos('helecho_julia.txt')  # Cambiar a 'helecho_bash.txt'

plt.figure(figsize=(6, 10))
plt.scatter(x_vals_bash, y_vals_bash, s=0.1, color='green')
plt.title("Helecho de Barnsley generado en Julia")
plt.axis('off')
plt.show()
````
