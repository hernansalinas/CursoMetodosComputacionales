```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
#from numpy.lib.npyio import savetxt
x0=0
v0=1
angulo=45
tetha=np.deg2rad(angulo)
#tiene que ser en radianes
global g
g=9.8
tv=(2*v0*np.sin(tetha))/g
t=np.linspace(0,tv,1000)
x=x0+v0*np.cos(tetha)*t
y=v0*np.sin(tetha)*t-((1/2)*g*(t**2))

fig, ax= plt.subplots()
ax.plot(x, y, color="pink")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

```python
X=np.array([x,y])
np.savetxt("Datos.txt",X.T)
```
