```python
import pandas as pd
import matplotlib.pyplot as plt
#se lee el archivo cvs
url="https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv"
x=pd.read_csv(url)
```

```python
#Se genera una nueva columna la cual trasforma los grados F a grados C
x["AvgTempCelcius"]=(x["AvgTemperature"]-32)/1.8
```

```python
#Se realiza el grafico de la temperatura de bogota con respecto a los años
bogota1 = x[x["City"] == "Bogota"]
plt.plot(bogota1["Year"], bogota1["AvgTempCelcius"],"b.-")
plt.grid(True)
```

```python
#Se realiza el grafico de la temperatura de barcelona con respecto a los años
Barcelona1 = x[x["City"] == "Barcelona"]
plt.plot(Barcelona1["Year"], Barcelona1["AvgTempCelcius"],"r.-")
plt.grid(True)
```

```python
#En esta celda se hallan las maximas temperaturas de ambas ciudades
maxtemp1 = x.loc[x["City"] == "Bogota", "AvgTempCelcius"].max()
maxtemp2 = x.loc[x["City"] == "Barcelona", "AvgTempCelcius"].max()
print("La maxima temperatura en celsius en la ciudad de Bogota fue: ",maxtemp1)
print("La maxima temperatura en celsius en la ciudad de Barcelona fue: ",maxtemp2)
```
