```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
```

```python
f=lambda x:x**2
```

```python
def integral(a,b,f,N):
  x=np.linspace(a,b,N)
  area=0
  for i in range(len(x)-1):
    deltax=x[i+1]-x[i]
    d=(f(x[i])+f(x[i+1]))/2
    deltaA=d*deltax
    area=area+deltaA
  return area
```

```python
integral(0,1,f,100000000)
```

```python
lin=np.linspace(0,1,100000000)
```

```python
plt.plot(lin,f(lin))
```

```python
sc.integrate.trapezoid(f(lin),lin)
```
