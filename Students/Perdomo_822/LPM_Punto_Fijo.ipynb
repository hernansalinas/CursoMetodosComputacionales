```python
import math as mp
import matplotlib.pyplot as plt
import numpy as np
import scipy as sc
import sympy as sp
```

```python
#Definición de funciones:

f = lambda x: (x-1)**2
g = lambda x: x-f(x)    #Se define g(x) según definición de punto fijo.

x=np.linspace(-2,2.5,100)

#gráfica:
plt.plot(x,f(x), label='f(x)')
plt.plot(x,g(x), label='g(x)')
plt.plot(x,x, label='X')
plt.title("Grafica de las funciones definidas")
plt.xlabel('X')
plt.legend()
plt.show()
```

Notese que el punto a encontrar es la intercección entre $Y(x)= x$ y la función $g(x)$, punto que será la raiz que buscamos de la función f(x).  

```python

def punto_fijo(x_0, f,N_Max):
  #Defino la funciòn donde x_0 es el punto inicial, f la función y N_Max el numero maximo de iteraciones del ciclo:
  g = lambda x: x-f(x)
  x0 = x_0

  #Iteraciones con ciclo while:
  xf = 0
  i = 0
  while abs(x0 - xf)>10e-9 and i < N_Max:
    xf = g(x0)
    x0 = g(xf)
    i += 1

  return (x0)


x=np.linspace(-1.5,2.5,100)

x0 = punto_fijo(1.6, f, 100)

#gráfica:
plt.plot(x,f(x), label='f(x)')
plt.plot(x,g(x), label='g(x)')
plt.plot(x,x, label='X')
plt.plot(x0,f(x0), '*', label='$(x_0,f(x_0))$')
plt.plot(x0,g(x0), '.', label='$(x_0,g(x_0))$')
plt.title("Grafica de las funciones y puntos encontrados")
plt.xlabel('X')
plt.legend()
plt.grid()
plt.show()



print('El punto es aproximadamente: ', x0)
print('El punto de intersección entre g(x) & X es: ', (x0,g(x0)))
print('El punto raiz de f(x) es: ', (x0,f(x0)))
```

Comparando con el punto fijo de la librería Scipy:
