\begin{equation}
m\frac{dv_x}{dt} = -mb(v_x-u\cos \alpha )

Posición en x
\begin{equation}
x=(u\cos \alpha) t + \frac{1}{b} (v_0\cos \theta_0 -u \cos\alpha)(1-\mathrm{exp}(-bt))
\end{equation}

Posicion en y
\begin{equation}
y= \frac{1}{b}\left(\frac{g}{b}+v_0 \sin \theta_0 -u\sin\alpha \right)(1-\mathrm{exp}(-bt))-\left( \frac{g}{b}-u\sin\alpha\right)t
\end{equation}

# 4. Fixed point methods.

When a new planet is discovered, there are different methods to estimate its physical properties. Many times is only possible to estimate either the planet mass or the planet radius and the other property has to be predicted through computer modelling.

If one has the planet mass, a very rough way to estimate its radius is to assume certain composition (mean density) and a homogeneous distribution (a very bad assumption!). For example, for the planet [Gliese 832c](http://es.wikipedia.org/wiki/Gliese_832_c) with a mass $M= 5.40 M_{\oplus}$, if we assume an earth-like composition, i.e. $\bar \rho_{\oplus} = 5520\ kg/m^3$, we obtain:

$$R_{g832c} = \left( \frac{3 M_{g832c}}{ 4 \pi \bar\rho_{\oplus} } \right)^{1/3} \approx 1.75 R_{\oplus}$$

That would be the planet radius if the composition where exactly equal to earth's.

A more realistic approach is assuming an internal one-layer density profile like:

$$\rho(r) = \rho_0 \exp\left( -\frac{r}{L} \right)$$

where $\rho_0$ is the density at planet centre and $L$ is a characteristic lenght depending on the composition. From numerical models of planet interiors, the estimated parameters for a planet of are $M= 5.40 M_{\oplus}$ are approximately $\rho_0 = 18000\ kg/m^3$ and $L = 6500\ km$.

Integrating over the planet volume, we obtain the total mass as

$$M = 4\pi \int_0^R \rho(r)r^2dr$$

This is a function of the mass in terms of the planet radius.

Solving the equation $M(R) = M_{g832c}$ it would be possible to find a more realistic planet radius. However when using numerical models, it is not possible to approach the solution from the left side as a negative mass makes no sense.

b). Deducción del metodo Newton- Raphson:

El método Newton-Raphson es un método para encontrar las raices de funciones reales; método que es útil a la hora de analizar funciones no lineales.  

El cual es aplicable cuando:

sea $f$ una función real derivable en un intervalo abierto que contiene una raiz $\alpha$, es decir, $f(\alpha) = 0$.

Por lo tanto para encontrar una aproximación de $\alpha$ usando este  método, se parte de un valor incial $x_n$ cercano a $\alpha$ y se aplica una expansión en *Serie de Taylor* de $f$ al rededor de un $X_n$. tal que:


  \begin{equation}
    f(x)= f(x_n) + f'(x_n)(x-x_n) + \frac{f''(\xi)}{2} (x-x_n)^2 + ...
  \end{equation}

Dónde $\xi$ es un punto entre $x$ y $x_n$. Si se hace $x= \alpha$, se obtiene:


  \begin{equation}
    0=f(\alpha)= f(x_n) + f'(x_n)(\alpha-x_n) + \frac{f''(\xi)}{2} (\alpha-x_n)^2 + ...
  \end{equation}

Que luego de algunas operaciones matemáticas se obtiene:

  \begin{equation}
    x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)} ,  \\ n= 1,2,3...
  \end{equation}


c). Tanto el método de Newton-Raphson y el método de la secante son dos algoritmos iterativos y se basan en la aproximación sucesiva de la raíz. sin embargo la diferencia entre estos métodos es que:

1. El método de Newton-Raphson utiliza la derivada de la función para mejorar la presicisión de la estimación de la raíz en cada iteración; pues en cada paso se aproxima la función por su recta tangente en un punto cercano a la raíz, sin importar si se falla con una estimación incial lejos de la raíz.

2. Mientras que el metodo de la secante aproxima a función por una recta que pasa por dos puntos cercanos a la raíz. Es decir se utilizan dos valores que estimamos inicialmente de la raíz.

3. Se puede decir que una diferencia más importante entre los dos métodos es que el de Newton-Raphson converge más rápiso que el método de la Secante.  

d). Pseudocódigo para aplicar el método Punto Fijo.

El método del punto fijo es un método iterativo que permite resolver sistemas de ecuaciones no necesariamente lineales. En particular, se puede utilizar para determinar raíces de una función de la forma f(x) = x. A continuación se presenta el pseudocódigo para el método del punto fijo:

```
Inicio
    Leer x0, f, N_Max
    Definir g(x) = x - f(x)
    x0 = x_0
    xf = 0
    i = 0
    Mientras |x0 - xf| > 10e-9 y i < N_Max hacer
        xf = g(x0)
        x0 = g(xf)
        i += 1
    Fin Mientras
    Escribir "La raíz es: ", x0
Fin

```

En este pseudocódigo se leen los valores iniciales de x0, f y N_Max. Luego, se define la función g(x) = x - f(x). Se establecen los valores iniciales de x0, xf e i. El ciclo Mientras se ejecuta mientras la diferencia entre x0 y xf sea mayor que epsilon(10e-9) y el número de iteraciones sea menor que N_Max. Dentro del ciclo, se actualiza el valor de xf mediante la función g(x0) y se actualiza el valor de x0 mediante la función g(xf). El ciclo continúa hasta que se alcance la tolerancia o el número máximo de iteraciones. Finalmente, se imprime el valor de la raíz.

````python
#Solución al punto 2 ceros de funciones:
import scipy as sc
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
#Para diccionarios:
import pandas as pd
````

````python
#Definiciòn de funciones:
f1 = lambda x: (np.exp(x)-2)
f2 = lambda x: np.cos((np.exp(x)-2))
f=  lambda x: f1(x)-f2(x)
````

````python
#Mètodo Secante:
x0= sc.optimize.newton(f,20)
print("La raiz es aproximadamente: ",x0)

````

````python
#Bisection: Miramos las gràficas de las funciones:
x=np.linspace(-1,3,1000)
plt.plot(x,f1(x), label='f1(x)=$e^x -1$')
plt.plot(x,f2(x), label= 'f2(x)= $\cos(f1)$')
plt.title('Gráfica de f1 y f2')
plt.xlabel('X')
plt.ylabel('f(x)')
plt.legend()
````

De la grafica notamos que la bisección entre las dos gráficas es aproximadamente 1. Por lo tanto, realizando calculo por librería, tenemos:

````python
#Bisection por libreria:
a=0
b=4
raiz = sc.optimize.bisect(f,a,b)
print('La raiz es aproximadamente: ', raiz)
````

````python
#Definimos la derivada para método Newton:
df = lambda x: np.exp(x) + np.sin(f1(x)*np.exp(x))
x1 = sc.optimize.newton(f,2,df)

print('La raiz por metodo N-R es aproximadamente: ',x1 )
````

````python

#Ecuaciones del Movimiento parabólico en presencia del viento:
v0 = 60                                       #Velocidad en m/s
alpha = np.pi                                 #Angulo del viento
theta = np.deg2rad(45)                        #Angulo de lanzamiento
u = v0/2                                      #rapidez del viento
b = 0.05                                      #Parametro de friccion
g = 9.8                                       #Aceleracion de la gravedad m/s2



# Defino funciones para posiciones:
def posxy (t, theta, alpha, v0, b, u, g):

  ca=u * np.cos(alpha)
  sa=u * np.sin(alpha)
  ct=v0 * np.cos(theta)
  st=v0 * np.sin(theta)
  e = 1 - np.exp(-b*t)

  X = ca*t + ((ct-ca)*e)/b

  Y= ((((g/b)+ st - sa)*e)/b) - (((g/b)-sa)*t)

  return X, Y

#Se determina un t aleatorio para buscarel valor aproximado a la raìz:
#t=np.linspace(0,8,300)

#Se utiliza el método de newton para encontrar el timepo de vuelo:
Y0= lambda x: (1 / b) * ((g / b) + (v0) * np.sin(theta) - u * np.sin(alpha)) * (1 - np.exp(-b * x)) - ((g / b) - u * np.sin(alpha)) * x
tv = sc.optimize.newton(Y0,8)

t = np.linspace(0,tv,1000)

#Gráfica
x,y = posxy(t,theta,alpha,v0,b,u,g)

plt.plot(x,y)
plt.title('X vs Y en Movimiento Parabólico Con Viento')
plt.xlabel('X(t)')
plt.ylabel('Y(t)')
plt.grid()
plt.show()

````

#5. Construcción de fractal:

````python
from PIL import Image

# Dimensiones de la imagen
Npx_X = 720
Npx_Y = 720

# Creación de la imagen en blanco
image = Image.new("RGB", (Npx_X, Npx_Y))

# Coordenadas en el plano complejo
xmin = -1
xmax = 1
ymin = -1
ymax = 1

# Mapeo de píxeles a coordenadas complejas
npx = np.arange(0, Npx_X)
npy = np.arange(0, Npx_Y)
zx = npx * (xmax - xmin) / (Npx_X - 1) + xmin
zy = npy * (ymax - ymin) / (Npx_Y - 1) + ymin

# Función y derivada
def f(z):
    return z**3 - 1

def fp(z):
    return 3 * z**2

# Parámetros
a = 0.2
Niteraciones = 120
eps = 10e-10

# Generación de la imagen
x = -1
for zx_ in zx:
    x = x + 1
    y = -1
    for zy_ in zy:
        y = y + 1
        Z_0 = complex(zx_, zy_)

        for i in range(0, Niteraciones):
            Z = Z_0 - a * f(Z_0) / fp(Z_0)
            if abs(Z - Z_0) <= eps:
                break

            Z_0 = Z

        r, g, b = i % 4 * 64, i % 8 * 16, i % 16 * 16
        image.putpixel((x, y), (r, g, b))

# Mostrar la imagen resultante
plt.imshow(image)
plt.show()
````


**Comprensión del código:**


Este código en Python genera una imagen gráfica utilizando la biblioteca **PIL** (Python Imaging Library). La imagen resultante es un fractal conocido como el conjunto de Mandelbrot.

donde los pasos que se hacen para realizar el fractal son los siguientes:

1. Primero, se importa la clase `Image` del módulo `PIL` para poder crear y manipular imágenes.
2. A continuación, se definen las dimensiones de la imagen en píxeles utilizando las variables `Npx_X` y `Npx_Y`.
3. Se crea una nueva imagen en blanco utilizando la función `Image.new()`, especificando el modo de color `"RGB"` y las dimensiones de la imagen.
4. Luego, se definen las coordenadas en el plano complejo utilizando las variables `xmin`, `xmax`, `ymin` y `ymax`. Estas coordenadas determinan el rango de valores que se mapearán a los píxeles de la imagen.
5. Se crea un mapeo de los píxeles a las coordenadas complejas utilizando las variables `npx`, `npy`, `zx` y `zy`. Estas variables se utilizan para calcular las coordenadas complejas correspondientes a cada píxel de la imagen.
6. A continuación, se definen dos funciones: `f(z)` y `fp(z)`. La función `f(z)` representa la función matemática utilizada para generar el fractal, mientras que la función `fp(z)` representa su derivada.
7. Se definen los parámetros del algoritmo, como el valor de `a`, el número máximo de iteraciones (`Niteraciones`) y un umbral de convergencia (`eps`).
8. Luego, se itera sobre cada píxel de la imagen utilizando dos bucles anidados. Para cada píxel, se inicializa una variable `Z_0` con las coordenadas complejas correspondientes al píxel.
9. A continuación, se realiza un bucle iterativo para calcular el valor de `Z` utilizando la fórmula iterativa del conjunto de Mandelbrot: `Z = Z_0 - a * f(Z_0) / fp(Z_0)`. El bucle continúa hasta que se alcanza un criterio de convergencia definido por el umbral `eps`.
10. Dentro del bucle iterativo, se actualiza el valor de `Z_0` con el valor calculado de `Z`.
11. Después de salir del bucle iterativo, se asigna un color al píxel en función del número de iteraciones realizadas (`i`). Los valores RGB del color se calculan utilizando operaciones matemáticas simples basadas en el número de iteraciones.
12. Finalmente, se muestra la imagen resultante utilizando la función `plt.imshow()` y se llama a la función `plt.show()` para mostrarla en una ventana separada.

Este código genera una imagen fractal interesante que muestra la estructura del conjunto de Mandelbrot. Cada color en la imagen representa una región diferente del conjunto fractal.
