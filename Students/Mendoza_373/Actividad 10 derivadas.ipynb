```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
f = lambda x: np.sin(x) + np.tan(x)**2
x1 = np.linspace(-0.1,0.1)

df = sc.misc.derivative(f,x1)

fp = lambda x: np.cos(x) + 2*np.tan(x) + 2*np.tan(x)**3
```

```python
plt.plot(x1,f(x1),label="f")
plt.plot(x1,df,label="f'")
plt.plot(x1,fp(x1))
plt.legend()
plt.grid()
plt.show()
```

```python
def derivada_num(f,X):

  h = 1e-10
  n = np.arange(0,len(X),1)

  df = np.zeros(len(X))

  for i in n:
     df[i] = (1/(2*h))*(f(X[i]+h)-f(X[i]-h))-(h**2)/(6)

  return df
```

```python
f = lambda x: np.sin(x) + np.tan(x)**2
fp = lambda x: np.cos(x) + 2*np.tan(x) + 2*np.tan(x)**3

x0 = np.linspace(-1,1)

Df = derivada_num(f,x0)

plt.plot(x0,f(x0))
plt.plot(x0,Df)
plt.plot(x0,fp(x0))
plt.grid()
plt.show()
```
