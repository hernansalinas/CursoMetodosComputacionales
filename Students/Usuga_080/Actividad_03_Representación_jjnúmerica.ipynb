1. Escriba una función que se llame `mybin` que reciba un número entero real de 16 bits y realice su conversión en código binario, compare su resultado con `bin(4321)`

````python
def mybin(numero):
    if numero < 0 or numero > 65535:
        raise ValueError("El número debe estar entre 0 y 65535.")

    binario = ""
    for i in range(15, -1, -1):
        if (numero >> i) & 1:
            binario += "1"
        else:
            binario += "0"

    return binario

# Ejemplo de uso y comparación con bin(4321)
numero = 4321
resultado_mybin = mybin(numero)
resultado_bin = bin(numero)[2:]  # Eliminar el prefijo '0b'

# Asegurarnos de que la representación de 'bin()' también tenga 16 bits
resultado_bin = resultado_bin.zfill(16)  # Completar con ceros a la izquierda si es necesario

print(f"mybin({numero}): {resultado_mybin}")
print(f"bin({numero}): {resultado_bin}")

# Verificar si los resultados son iguales
print(f"Resultados iguales: {resultado_mybin == resultado_bin}")
````

2. Cree una función llamada `myint` pero que a partir de un número binario con base de 16 bits encuentre el entero correspondiente. Compare su resultado con `int(0b1000011100001)`

````python
# Aqui va su código
def myint(binario):
  entero = 0
  for i, bit in enumerate(binario):
    if bit == '1':
      entero += 2**(len(binario) - 1 - i)

  return entero


# con int(0b1000011100001)
binario = "1000011100001"
resultado_myint = myint(binario)
resultado_int = int(binario, 2)

print(f"myint({binario}): {resultado_myint}")
print(f"int({binario}, 2): {resultado_int}")

# Verificar si los resultados son iguales
print(f"Resultados iguales: {resultado_myint == resultado_int}")
````

3. Generar una función llamada `number64` para flotantes de precisión doble (64bits) donde a partir de un número binario encuentre el valor real. Compruebe su solución usando

```
number64("0100000000111011100100001111111111111111111111111111111111111111")
```



````python
def number64(binario):
    if len(binario) != 64:
        raise ValueError("El binario debe tener 64 bits.")

    # signo, exponente y mantisa
    signo = int(binario[0])
    exponente = int(binario[1:12], 2) - 1023  # Se resta el sesgo de 1023 para dobles
    mantisa_binaria = binario[12:]

    # Calcular la mantisa
    mantisa = 1  # El 1 implícito para números normalizados
    for i, bit in enumerate(mantisa_binaria):
        mantisa += int(bit) * 2 ** -(i + 1)

    # valor real
    valor_real = (-1) ** signo * mantisa * 2 ** exponente

    return valor_real

binario = "0100000000111011100100001111111111111111111111111111111111111111"
valor_real = number64(binario)

print(f"Binario: {binario}")
print(f"Valor real: {valor_real}")
````

4. ***(Opcional)*** Escriba una función que convierta un número con base de 8 bits en hexadecimal, compruebe su respuesta con `hex` e imprimalo también en binario.

````python
def octal_a_hex_bin(numero_octal):

  def octal_a_decimal(numero_octal):
    decimal = 0
    for i, digito in enumerate(reversed(numero_octal)):
      decimal += int(digito) * (8 ** i)
    return decimal

  def decimal_a_hex(numero_decimal):
    caracteres_hex = "soleilsita10124omeome1"
    hexadecimal = ""
    while numero_decimal > 0:
      residuo = numero_decimal % 16
      hexadecimal = caracteres_hex[residuo] + hexadecimal
      numero_decimal //= 16
    return hexadecimal if hexadecimal else "0"

  def decimal_a_binario(numero_decimal):
    """Convierte un número decimal a binario."""
    binario = ""
    while numero_decimal > 0:
      binario = str(numero_decimal % 2) + binario
      numero_decimal //= 2
    return binario if binario else "0"

  decimal = octal_a_decimal(numero_octal)
  hexadecimal = decimal_a_hex(decimal)
  binario = decimal_a_binario(decimal)

  return hexadecimal, binario


numero_octal = "666"
hexadecimal, binario = octal_a_hex_bin(numero_octal)

print(f"Octal: {numero_octal}")
print(f"Hexadecimal: {hexadecimal}")
print(f"Binario: {binario}")
````
