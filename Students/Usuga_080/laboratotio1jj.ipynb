```python
import random

def adivina():
    num_secr = random.randint(1, 100)
    print("Adivina el Número")
    print("Adivina un número entre 1 y 100")

    inten = 0
    adivinado = False

    while not adivinado:
        try:
            suposicion = int(input("Introduce tu suposición: "))
        except ValueError:
            print("Por favor, introduce un número válido.")
            continue

        inten += 1

        if suposicion < num_secr:
            print("Cuchito, muy bajo. ¡Intenta de nuevo!")
        elif suposicion > num_secr:
            print("Pai, muy alto. ¡Intenta de nuevo!")
        else:
            print(f"¡Por fin adivinaste después de {inten} intentos!")
            adivinado = True

    print("Gracias por jugar. ¡Hasta la próxima!")

# Ejecutar el juego
adivina()

```
