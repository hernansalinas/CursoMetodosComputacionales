### Factorial
1. Elaborar un algoritmo e implementar en python, el factorial de un número entero `n` ingresado por el usuario.

1. Si el número es un entero deberá retornal el factorial del número.
2. Si el número es negativo deberá aparecer un mensaje tipo string con el mensaje  "el número no puede ser negativo"

3. Si el número es pertenece a los reales(float) el mensaje será, el número no puede ser real.


     ### Ejemplo de Ejecución
    ```python

    >>> factorial(3)
        6

    >>> factorial(1987123)
        =???
```


````python
def factorial(n):
    if isinstance(n, str):
        return "Debe ser un número entero positivo"

    if not isinstance(n, (int, float)):
        return "Debe ser un número entero positivo"

    if isinstance(n, float):
        return "El número no puede ser real"

    if n < 0:
        return "El número no puede ser negativo"

    if isinstance(n, int):
        factorial = 1
        for i in range(1, n + 1):
            factorial *= i
        return factorial
````

Problema no obligatorio:

P1. Retornar los 30 primeros números de cada clase

````python
def divisores_propios(n):
    return [i for i in range(1, n) if n % i == 0]
def numeros_defectivos(n):
    return sum(divisores_propios(n)) < n
def numeros_abundantes(n):
    return sum(divisores_propios(n)) > n
````

````python
from itertools import combinations

def numeros_semiperfectos(n):
    divisores = divisores_propios(n)
    # Probar todas las combinaciones de divisores
    for i in range(1, len(divisores) + 1):
        for combo in combinations(divisores, i):
            if sum(combo) == n:
                return True
    return False
````

````python
def numeros_perfectos(n):
    return sum(divisores_propios(n)) == n
````

````python
def numeros_primos(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
````

````python
print(numeros_defectivos(30))  # True, porque 1 + 2 + 4 = 7 < 8
print(numeros_abundantes(30))  # True, porque 1 + 2 + 3 + 4 + 6 = 16 > 12
print(numeros_semiperfectos(30))  # True, porque 6 + 4 + 2 = 12
print(numeros_perfectos(30))  # True, porque 1 + 2 + 3 = 6
print(numeros_primos(30))  # True, porque 11 es divisible solo por 1 y 11
````


7. Diseñar un programa en el que entrado dos números `a`  y `b` retorne una variable booleana **True** o **false**.

  [Números amigos](https://es.wikipedia.org/wiki/N%C3%BAmeros_amigos) `a` y `b` tales que a es la suma de los divisores propios de `b` y viceversa.
    La rutina se debera llamar numeros_amigos




P2. Retornar los 10 primeros pares de numeros perfectos, semiperfectos, amigos



````python
def numeros_amigos (a,b):
  div_propios_a, div_propios_b = [], []
  for i in range (1, a):
    if a%i == 0:
      div_propios_a.append(i)
  sum_a = int(sum(div_propios_a))
  for i in range (1, b):
    if b%i == 0:
      div_propios_b.append(i)
  sum_b = int(sum(div_propios_b))

  if a == sum_b and b == sum_a:
    return True
  else:
    return False
````

3. Determine si un número `n` entero ingresado por el usuario es un [palíndromo](https://en.wikipedia.org/wiki/Palindromic_number), (Retorne `True` en caso afirmativo y `False` en caso contrario)

```python


    >>> palindromo(3333333)
        True

    >>> palindromo(2323)
        True
        
    >>> palindromo(1111349111111)
        False
```




````python
def palindromo(n):
    n_str = str(n)
    return n_str == n_str[::-1]
print(palindromo(3333333))
print(palindromo(2323))
print(palindromo(1111349111111))
````

Sólo para comprobar tu código,  puedes hacer uso del comando sort de python.
```python
<<< b = [22, 32, 42, 12, 22, 31, 41, 11, 12, 232, 24, 12, 22]

<<< print(b.sort())

<<< [11, 12, 12, 12, 22, 22, 22, 24, 31, 32, 41, 42, 232]
```

````python
def quicksort(arr):
    if len(arr) < 2:
        return arr  # Base case: already sorted
    else:
        pivot = arr[0]  # Choose the first element as the pivot
        less = [i for i in arr[1:] if i <= pivot]  # Elements less than or equal to the pivot
        greater = [i for i in arr[1:] if i > pivot]  # Elements greater than the pivot
        return quicksort(less) + [pivot] + quicksort(greater)  # Recursively sort and combine

# Example usage:
v = [22, 32, 42, 12, 22, 31, 41, 11, 12, 232, 24, 12, 22]
sorted_v = quicksort(v)
print(sorted_v)  # Output: [11, 12, 12, 12, 22, 22, 22, 24, 31, 32, 41, 42, 232]
````
