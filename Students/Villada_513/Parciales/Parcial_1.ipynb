````python
g  = 9.81
l = 4
t = 2.5

v = lambda H: np.sqrt(2*g*H) * np.tanh((np.sqrt(2*g*H) / 2*l) * t)

h_vals = np.linspace(0, 5, 100)
v_vals = v(h_vals)

def metodo_biseccion(f, a, b, tol):
  while (b - a)/2 > tol :

      c = (a + b)/2

      if f(c) == 0 :
          return c

      if f(a)*f(c) < 0 :
            b = c
      else:
            a = c

  return(a + b)/2

H = sp.symbols("H")
v_newton = sp.sqrt(2*g*H) * sp.tanh((sp.sqrt(2*g*H) / 2*l) * t)
vd = sp.diff(v_newton, H)
vd = sp.lambdify(H, vd, "numpy")

v = lambda H: np.sqrt(2*g*H) * np.tanh((np.sqrt(2*g*H) / 2*l) * t) - 5



print("Raiz biseccion", metodo_biseccion(v, 0, 5, 1e-7))
print("Raiz Newton", newton(v, 3, vd))

plt.plot(h_vals, v_vals)
plt.xlabel("H")
plt.ylabel("v")
plt.axvline(newton(v, 3, vd))
plt.grid(True)
````
