```python
import numpy as np
import matplotlib.pylab as plt
```

```python
puntos_x=np.array([1,2,3])
puntos_y=np.array([2,8,1])
plt.figure()
plt.plot(puntos_x,puntos_y,"ro")
plt.show()
```

```python
#recurrencia de la pendiente
m=(puntos_y[1]-puntos_y[0])/(puntos_x[1]-puntos_x[0])
print(m)
```

```python
m=(puntos_y[2]-puntos_y[1])/(puntos_x[2]-puntos_x[1])
print(m)
```

```python
#trazar la recta para unir dos puntos
m=(puntos_y[1]-puntos_y[0])/(puntos_x[1]-puntos_x[0])
x_=np.linspace(puntos_x[0],puntos_x[1],100)
y_=puntos_y[0]+m*(x_-puntos_x[0])
plt.figure()
plt.plot(puntos_x,puntos_y,"ro")
plt.plot(x_,y_)
plt.grid(True)
plt.show()
```

```python
#trazar la recta para unir los puntos

#Valores de las pendientes
m1=(puntos_y[1]-puntos_y[0])/(puntos_x[1]-puntos_x[0])
m2=(puntos_y[2]-puntos_y[1])/(puntos_x[2]-puntos_x[1])

#primera recta
x1_=np.linspace(puntos_x[0],puntos_x[1],100)
y1_=puntos_y[0]+m1*(x1_-puntos_x[0])

#segunda recta
x2_=np.linspace(puntos_x[1],puntos_x[2],100)
y2_=puntos_y[1]+m2*(x2_-puntos_x[1])

#Gr√°fico
plt.figure()
plt.plot(puntos_x,puntos_y,"ro")
plt.plot(x1_,y1_)
plt.plot(x2_,y2_)
plt.grid(True)
plt.show()
```

```python
#Generalizaci√≥n del c√≥digo
N=int(input("ingrese cuantos puntos desea graficar:"))
if N<2:
    print("Se requieren al menos dos puntos para calcular la pendiente ")
else:
    puntos_x=np.zeros(N)
    puntos_y=np.zeros(N)
    for i in range(N):
        x=float(input(f"ingrese la coordenada x del punto{i+1}:"))
        y=float(input(f"ingrese la coordenada y del punto{i+1}:"))
        puntos_x[i]=x
        puntos_y[i]=y

#C√°lculo de las pendientes y la muestra de coordenadas
for i in range(N-1):
    m=(puntos_y[i+1]-puntos_y[i])/(puntos_x[i+1]-puntos_x[i])
    print(f"pendientente entre ({puntos_x[i]},{puntos_y[i]}y ({puntos_x[i+1]},{puntos_y[i+1]}):{m}")

#Gr√°fico
plt.plot(puntos_x,puntos_y,"o-")
plt.xlabel("coordenada x")
plt.ylabel("coordenada y")
plt.title("Gr√°fica de los puntos ingresados")
plt.grid(True)
plt.show()

```
