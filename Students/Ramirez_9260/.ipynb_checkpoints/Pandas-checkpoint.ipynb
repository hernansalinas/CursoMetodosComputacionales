```python
import pandas as pd
import matplotlib.pyplot as plt


url="https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv"
```

```python
from tables.table import Column
df=pd.read_csv(url)
df=df.rename(columns={"Año":"Ano"})
df=df.drop(columns=["Unnamed: 0"])
df["AvgTemperatureCelsius"]=(df.AvgTemperature-32)/1.8
tf=df.City=="Bogota"
tf
x = df.Year
y = df.AvgTemperatureCelsius
plt.plot(x,y)
```

```python
df1 = df[df["City"] == "Bogota" ]
df2 = df[df["City"] == "Barcelona" ]
```

```python
x = df1.Year
y = df1.AvgTemperatureCelsius
plt.xlabel('Años')
plt.ylabel('Temperatura')
plt.title('Temperatura en el tiempo de Bogota')
plt.plot(x,y)
```

```python
x = df2.Year
y = df2.AvgTemperatureCelsius
plt.xlabel('Años')
plt.ylabel('Temperatura')
plt.title('Temperatura en el tiempo de Barcelona')
plt.plot(x,y)
```
