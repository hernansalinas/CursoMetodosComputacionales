```python
import numpy as np
import matplotlib.pyplot as plt

def calcular_campo_electrico_total(cargas, x, y):
    k = 9 * 10**9
    Ex_total = 0
    Ey_total = 0

    for carga in cargas:
        q, rx, ry = carga
        r = np.sqrt((x - rx)**2 + (y - ry)**2)
        Ex = (k * q * (x - rx)) / r**3
        Ey = (k * q * (y - ry)) / r**3
        Ex_total += Ex
        Ey_total += Ey

    return Ex_total, Ey_total

# Ingresar los valores de las cargas
q1 = float(input("Digite el valor de la carga 1: "))
q2 = float(input("Digite el valor de la carga 2: "))
q3 = float(input("Digite el valor de la carga 3: "))

# Datos de las cargas y posiciones
cargas = [
    (q1, -3, -2),
    (q2, 0, 2),
    (q3, 4, 0)
]

plt.figure()

# Calcular y graficar el campo elctrico en todos los puntos
x_range = np.linspace(-10, 10, 20)
y_range = np.linspace(-10, 10, 20)

for x in x_range:
    for y in y_range:
        Ex_total, Ey_total = calcular_campo_electrico_total(cargas, x, y)
        norm_factor = np.sqrt(Ex_total**2 + Ey_total**2)
        plt.quiver(x, y, Ex_total/norm_factor, Ey_total/norm_factor, angles='xy', scale_units='xy', scale=1.5, color='blue')

# Graficar las posiciones de las cargas
for carga in cargas:
    q, rx, ry = carga
    plt.scatter(rx, ry, color='red', s=100)

#plt.xlim(-10, 10)
#plt.ylim(-10, 10)
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.title('Campo electrico en todos los puntos para 3 cargas')
plt.grid()
plt.show()
```
