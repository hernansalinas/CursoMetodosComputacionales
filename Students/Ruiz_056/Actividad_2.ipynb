`````python
import pandas as pd

# Dataset 1
# URL of the webpage
url = "https://worldpopulationreview.com/country-rankings/average-height-by-country"

# Read the HTML content
html_content = pd.read_html(url)

alturas = html_content[2]

# Dataset 2
# URL of the webpage
url = "https://worldpopulationreview.com/country-rankings/average-net-worth-by-country"

# Read the HTML content
html_content = pd.read_html(url)

income = html_content[0]
alturas
`````

`````python
#a)
alturas=alturas.drop(columns=['Boys 19 (in)',"Girls 19 (in)"])  #Eliminar las columnas que están en "in"
alturas.rename(columns={'Average Height, Boys, 19 yo (cm)': 'Boys 19 (cm)'}, inplace=True) #Cambié el nombre de la segunda columna porque estaba muy distinta al resto
`````

`````python
alturas.columns = [col.replace("Boys", "Niño").replace("Girls", "Niña").replace("Country","País") for col in alturas.columns]
#Cambiar los nombres de las columnas al español
alturas
`````

`````python
#b)
alturas=alturas.drop(alturas.index[-1])  #eliminamos la ultima fila de la tabla que corresponde a la suma de todos los datos
alturas_1=alturas.head(10)  #Llamamos las primeras 10 filas de la tabla
fila_a_agregar = alturas.loc[alturas['País'] == 'Colombia']
alturas_1 = pd.concat([alturas_1, fila_a_agregar], ignore_index=True)  #Agregamos los datos de Colombia
alturas_1 = pd.concat([alturas_1, alturas.tail(10).head(10)], ignore_index=True)  #Agregamos las últimas 10 filas (paises con menor altura)
alturas_1
`````

`````python
#Sacamos los promedios
niño19=alturas_1['Niño 19 (cm)'].mean()
niña19=alturas_1['Niña 19 (cm)'].mean()

niño15=alturas_1['Niño 15 (cm)'].mean()
niña15=alturas_1['Niña 15 (cm)'].mean()

niño10=alturas_1['Niño 10 (cm)'].mean()
niña10=alturas_1['Niña 10 (cm)'].mean()

niño5=alturas_1['Niño 5 (cm)'].mean()
niña5=alturas_1['Niña 5 (cm)'].mean()

data1={                       #Creamos una tabla com las alturas promedio
    "19 años":[niño19,niña19],
    "15 años":[niño15,niña15],
    "10 años":[niño10,niña10],
    "5 años":[niño5,niña5]
}

tabla1=pd.DataFrame(data1)
tabla1=tabla1.transpose()   #Traspuse la tabla porque me estaba dando al revés
tabla1.columns=["Niños", "Niñas"]
tabla1.plot(kind = 'bar')
tabla1
`````

`````python
#c)
alturas_19=alturas_1.iloc[:, :3]  #10 mayores y 10 menores alturas para niños y niñas de 19 años
alturas_19
`````

`````python
income.columns=[col.replace("Country", "País") for col in income.columns]   #cambiamos el nombre de la primera columna por "País"
`````

`````python
tabla_final=pd.merge(alturas_19, income[["País","GDP per Adult (USD)"]], on='País', how='inner')
tabla_final.columns = [col.replace("GDP per Adult (USD)", "GDP") for col in tabla_final.columns]  #A la columna de GDP le puse un nombre más corto
tabla_final   #Tabla con las alturas de niños y niñas de 19 años y el GDP correspondiente de cada país
`````

`````python
#Cambiamos las comas por puntos, quitamos el signo $ y luego convertimos a float la columna de GDP
tabla_final['GDP'] = tabla_final['GDP'].str.replace(',', '').str.replace('$', '').astype(float)
tabla_final['GDP'] = pd.to_numeric(tabla_final['GDP'], errors='coerce')
tabla_final['Niño 19 (cm)'] = pd.to_numeric(tabla_final['Niño 19 (cm)'], errors='coerce')  #Cambiamos los valores tipo string a flotantes
tabla_final
`````

`````python
#Creamos las gráficas
#En esta miramos la altura promedio de niños de 19 años por país
plt.bar(tabla_final['País'], tabla_final['Niño 19 (cm)'], color='skyblue')
plt.xlabel('País')
plt.ylabel('Altura Promedio (cm)')
plt.title('Altura Promedio niños de 19 años por País')
plt.xticks(rotation=90)
plt.show()
`````

`````python
#En esta miramos la altura promedio de niñas de 19 años por país
plt.bar(tabla_final['País'], tabla_final['Niña 19 (cm)'], color='pink')
plt.xlabel('País')
plt.ylabel('Altura Promedio (cm)')
plt.title('Altura Promedio niñas de 19 años por País')
plt.xticks(rotation=90)
plt.show()
`````

`````python
#En esta relacionamos el GDP de cada país
plt.bar(tabla_final['País'], tabla_final['GDP'], color='orange')
plt.xlabel('País')
plt.ylabel('GDP')
plt.title('GDP por país')
plt.xticks(rotation=90)
plt.show()
`````
