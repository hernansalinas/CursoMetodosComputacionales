### Factorial
1. Elaborar un algoritmo e implementar en python, el factorial de un número entero `n` ingresado por el usuario.

1. Si el número es un entero deberá retornal el factorial del número.
2. Si el número es negativo deberá aparecer un mensaje tipo string con el mensaje  "el número no puede ser negativo"

3. Si el número es pertenece a los reales(float) el mensaje será, el número no puede ser real.


     ### Ejemplo de Ejecución
    ```python

    >>> factorial(3)
        6

    >>> factorial(1987123)
        =???
```


````python
#factorial de un numero
def factorial(n):
  fact = 1
  if n < 0:
    return ("el numero no puede ser negativo")
  elif type(n) == float:
    return ("el numero no puede ser real")
  else:
    for i in range(1,n+1):
      fact *= i
  return fact

n=int(input())
factorial(n)
````

````python
#numeros defectivos
divisores = []
def numeros_defectivos(n):
  sum = 0
  for i in range(1,n):
    if n % i == 0:
      sum = sum + i
      divisores.append(i)
  if sum < n:
    return True
  else:
    return False

numeros_defectivos(12)
````

````python
#numeros semiperfectos
from itertools import combinations
````

````python
div = []
comb = []
def numeros_semiperfectos(n):
  for i in range(1,n):
    if n%i == 0:
      div.append(i)
  for i in range(2, len(div)+1):
    com = list(combinations(div,i))
    comb.append(com)

  for i in range(len(comb)):
    for j in range(len(comb[i][:])):
      if sum(comb[i][j]) == n:
        return True
  else:
    return False

numeros_semiperfectos(21)
````

P2. Retornar los 10 primeros pares de numeros perfectos, semiperfectos, amigos



````python
#numeros amigos
div_a = []
div_b = []
def numeros_amigos(a,b):
  sum_a = 0
  sum_b = 0
  for i in range(1,a):
    if a % i == 0:
      sum_a = sum_a +i
      div_a.append(i)
  for i in range(1,b):
    if b % i == 0:
      sum_b = sum_b + i
      div_b.append(i)
  if sum_b == a and sum_a == b:
    return True
  else:
    return False

numeros_amigos(1184,1210)
````

3. Determine si un número `n` entero ingresado por el usuario es un [palíndromo](https://en.wikipedia.org/wiki/Palindromic_number), (Retorne `True` en caso afirmativo y `False` en caso contrario)

```python


    >>> palindromo(3333333)
        True

    >>> palindromo(2323)
        True
        
    >>> palindromo(1111349111111)
        False
```




````python
#numero palindromo
def palindromo(b):
  b_str = str(b)
  if b_str == b_str[::-1]:
    return True
  else:
    return False

b = int(input())
palindromo(b)
````

````python
#lista ordenada de menor a mayor
a = [8,22,9,44,55,23,47,469,1,5,22]
def quicksort(a):
  lista_ordenada = 0
  for i in range(0,len(a)):
    for j in range(i+1,len(a)):
      if a[i] > a[j]:
        lista_ordenada = a[i]
        a[i] = a[j]
        a[j] = lista_ordenada
  return a

b = print(quicksort(a))
````
