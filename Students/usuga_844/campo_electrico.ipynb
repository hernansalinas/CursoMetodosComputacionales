```python

# librerias
import numpy as np
import matplotlib.pylab as plt
```

```python
def campo_electrico(xp, yp, x1, y1, q1):
  """
  Esta funcion permite determinar el campo electrico en un punto
  xp :
  yp :
  """
  DeltaX = xp -x1
  DeltaY = yp -y1
  Deltar = np.sqrt(DeltaX**2+DeltaY**2)
  ux = DeltaX/Deltar
  uy = DeltaY/Deltar

  Ex = (k*q1/Deltar**2)*ux
  Ey = (k*q1/Deltar**2)*uy
  E=np.array([Ex, Ey])
  return E
# contante de k
k= 9*10**-9
# Carga 1
#Posicion y manitud de la carga
q1 = 2 # nano Coulomb
x1 = -1
y1 = 0
#Carga 1
#Posicion y manitud de la carga
q2 = -2 # nano Coulomb
x2 = 1
y2 = 0
```

```python
#Punto evaluacion
xp1= 0.5
yp1 = 0.5
E1=campo_electrico(xp1, yp1, x1, y1, q1)
E2=campo_electrico(xp1, yp1, x2, y2, q2)
E11=E1+E2
plt.figure()
plt.plot(x1, y1,"ro" )
plt.plot(x2, y2,"ro" )
plt.plot(xp1, yp1,"bo" )
plt.quiver(xp1, yp1, E11[0], E11[1])




plt.ylim(-0.1, 1.5)
plt.show()
```

```python
#campo electrico de un dipolo
def campo_electrico(xp, yp, x1, y1, q1):
    k = 9 * 10**9  # Constante de Coulomb
    DeltaX = xp - x1
    DeltaY = yp - y1
    Deltar = np.sqrt(DeltaX**2 + DeltaY**2)
    ux = DeltaX / Deltar
    uy = DeltaY / Deltar

    Ex = (k * q1 / Deltar**2) * ux
    Ey = (k * q1 / Deltar**2) * uy
    E = np.array([Ex, Ey])
    return E

# Constante de k
k = 9 * 10**9

# Carga 1
q1 = 2 * 10**-9  # nano Coulomb
x1 = -1
y1 = 1

# Carga 2
q2 = -2 * 10**-9  # nano Coulomb
x2 = 1
y2 = -1

# Puntos para evaluacin del campo elctrico
num_points = 10
x_vals = np.linspace(-2, 2, num_points)
y_vals = np.linspace(-1, 1, num_points)

Ex_total = np.zeros((num_points, num_points))
Ey_total = np.zeros((num_points, num_points))

for x_idx in range(num_points):
    for y_idx in range(num_points):
        xp = x_vals[x_idx]
        yp = y_vals[y_idx]
        E1 = campo_electrico(xp, yp, x1, y1, q1)
        E2 = campo_electrico(xp, yp, x2, y2, q2)
        Ex_total[x_idx, y_idx] = E1[0] + E2[0]
        Ey_total[x_idx, y_idx] = E1[1] + E2[1]
        magnitud = np.sqrt(Ex_total[x_idx, y_idx]**2 + Ey_total[x_idx, y_idx]**2)
        Ex_total[x_idx, y_idx] /= magnitud
        Ey_total[x_idx, y_idx] /= magnitud

plt.figure()
plt.plot(x1, y1, "ro")
plt.plot(x2, y2, "ro")
plt.plot(x_vals, y_vals, "bo")
plt.quiver(x_vals, y_vals, Ex_total, Ey_total, scale=20)
plt.ylim(-1.2, 1.2)
plt.xlim(-2.2, 2.2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Campo Elctrico")
plt.show()
```
