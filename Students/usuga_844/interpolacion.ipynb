```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python

x= np.array([1,2,3,4,5])
y= np.array([4,5,3,1,2])

```

```python
plt.figure()
plt.plot(x[0],y[0],"o")
plt.plot(x[1],y[1],"o")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python

m= (y[1]-y[0])/(x[1]-x[0])
m
x_=np.linspace(x[0],x[1],100)
y_= y[0]+ m * (x_-x[0])

plt.figure()
plt.plot(x,y,"o")
plt.plot(x_,y_)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
xx=np.zeros((len(x) - 1, 100))
yy=np.zeros((len(x) - 1, 100))


for i in range(len(x) - 1):
    m = (y[i + 1] - y[i]) / (x[i + 1] - x[i])
    x_ = np.linspace(x[i], x[i + 1], 100)
    y_ = y[i] + m * (x_ - x[i])
    xx[i]=x_
    yy[i]=y_



plt.figure()
for i in range(0,len(x) - 1):

  plt.plot(xx[i],yy[i])

plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python

x= np.array([1,2,3,4,5])
y= np.array([4,5,3,1,2])

x_=np.linspace(1,5,100)
y_=interpolate.interp1d(x,y)

plt.figure()
plt.plot (x ,y ,'o', x_ , y_ ( x_ ) ,'-')
plt.title (' Interpolacin lineal')
plt.xlabel("x")
plt.ylabel("y")
plt.grid ()
```
