# Modelo de Presion

Construir un grafico que muestre la variacion de presion de profundidad


```python
r = 997 # Densidad del agua
g = 9.8 # Gravedad
pa = 1013 # Presion atmosferica
presion = lambda y: pa + r*g*y

X = np.linspace(0,10,10) # 10 metros de produndidad maxima
Y = presion(X)


plt.plot(X, Y, "o-", label = f"{Y[-1]} Pa en profundidad maxima")
plt.ylabel("Presion (Pa)", fontsize = 15)
plt.xlabel("Profundidad (mts)", fontsize = 15)
plt.xticks(fontsize = 14)
plt.legend()
```

# Modelo de tiro parabolico

Dado un tetha y una velocidad inicial encestara (baloncesto)

```python
def resolver_formula_cuadratica(a, b, c):
    # Calcular el discriminante
    discriminante = math.sqrt(b**2 - 4*a*c)
    # Calcular las soluciones
    solucion1 = (-b + discriminante) / (2*a)
    solucion2 = (-b - discriminante) / (2*a)
    return solucion1, solucion2
```

```python
# Calculo de la Trayectoria
def movimiento_parabolico(vo,d,p,angulo):
  theta_grados = angulo
  theta = np.deg2rad(theta_grados)
  g = 9.8 # Gravedad de la  tierra
  trayectoriaX = lambda vox, t: vox* t
  trayectoriaY = lambda voy, g, t, p: (p + (voy*t))- (0.5*g*(t**2))
  vox = (vo * (np.cos(theta))) # Velocidad en x
  voy = (vo * (np.sin(theta))) # Velocidad en y
  hmax = p + ((voy**2) / (2*g)) # Altura maxima
  t_c = d / (vox) # Tiempo hasta d
  if t_c == 0: # En caso que se dispara desde abajo del aro
    X1 = lambda vox, t: 0* t
    a,b = resolver_formula_cuadratica((-0.5*g),voy,p)
    if isinstance(a, float) and a > 0:
      t_c = a
      tiempo = np.linspace(0, t_c, 50)
      hmax = p + ((voy**2) / (2*g))
      Y = trayectoriaY(vo,g,tiempo, p)
      X = X1(vox, tiempo)
      hmax = (vo**2) / (2*g)
      return X,Y, hmax
    elif isinstance(b, float) and b > 0:
      t_c = b
      tiempo = np.linspace(0, t_c, 50)
      Y = trayectoriaY(vo,g,tiempo, p)
      X = X1(vox,tiempo)
      hmax = (vo**2) / (2*g)
      return X,Y,hmax
  else:
    tiempo = np.linspace(0, t_c, 50)
    X = trayectoriaX(vox,tiempo)
    Y = trayectoriaY(voy,g,tiempo, p)
    return X,Y, hmax
```

```python
def tiro(X,Y):
  ty = False
  tx = False
  coordenadas = tuple(zip(X, Y))
  for e in coordenadas:
    x,y = e # con esto se toda la trayectoria en x,y
    tx = math.isclose(d,x,abs_tol=0.45)  # math.isclose retorna True si se cumple la condicin
    ty = math.isclose(3.05,y,abs_tol = 0.45) # el ancho del aro es 450 mm == 0.45 mts, 3.05 la altura de la cesta
    if ty == True and tx == True:
      return print("encesto")
  if ty == False or tx == False:
    return print("No encesto")
```

Leer el arhivo json

```python
with open("input.json", "r") as f:
  data = json.load(f)
angulo = data["angulo"] # Angulo conforme a la horizontal
vo = data["vo"] # Velocidad inicial
p = data["p"] # Altura de la persona
d = data["d"] # Distancia hasta el arco
print(angulo, vo,p, d)
```

{"d": 10, "angulo": 32, "vo": 12, "p": 1.8} para esstos datos encesta

```python
X_,Y_,hr = movimiento_parabolico(vo,d,p,angulo)
tiro(X_,Y_)
```

```python
# Grafica
if all(elemento == 0 for elemento in X_): # Si se dispara bajo el aro
  X_ = [x + 1 for x in X_] # No grafica en el origen, no se como, asi que hice esto para que lo graficara
  plt.plot(X_,Y_)
  plt.xlabel("x(mts)", fontsize = 15)
  plt.ylabel("y(mts)", fontsize = 15)
  plt.xlim(0, 2)
  plt.ylim(0, (hr + 1))
  plt.show()
else:
  plt.plot(X_,Y_)
  plt.xlabel("x(mts)", fontsize = 15)
  plt.ylabel("y(mts)", fontsize = 15)
  plt.xlim(0, d)
  plt.ylim(0, (hr + 1))
  plt.show()
```
