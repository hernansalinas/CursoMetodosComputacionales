```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url='https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv'
```

```python
df=pd.read_csv(url)
df
```

```python
df['celcius']= (df.AvgTemperature-32/1.8) #nueva columna
```

```python
tf=df['City']=='Bogota'
df1=df[tf].copy()
df1
```

```python
#grfico de AvgTempCelcius como funcin del ao para la ciudad de Bogot
grados=df1.celcius.unique()
aos=df1.Year.unique()
plt.figure()
plt.plot(aos,grados, '.g-')
plt.title('Temperatura Bogot', fontsize=30)
plt.xlabel('Years', fontsize=30)
plt.ylabel('Celcius', fontsize=30)
plt.grid()
```

```python
tf1=df['City']=='Barcelona'
df2=df[tf1].copy()
df2
```

```python
#grfico de AvgTempCelcius como funcin del ao para la ciudad de Barcelona
grados=df2.celcius.unique()
aos=df2.Year.unique()
plt.figure()
plt.plot(aos,grados, '.g-')
plt.title('Temperatura Barcelona', fontsize=30)
plt.xlabel('Aos', fontsize=25)
plt.ylabel('Celsius', fontsize=25)
plt.grid()
```
