1. Escriba una función que se llame `mybin` que reciba un número entero real de 16 bits y realice su conversión en código binario, compare su resultado con `bin(4321)`

````python
# Aqui va su código

def mybin(n: int) -> str:

  """
  La función retorna la representación binaria de un número entero n.
  """

  if (2 ** 16 - 1) > n:

    binario = '0b'
    while n > 0:
      binario += str(n % 2)
      n //= 2

    return binario

  else:
    return f'El número {n} requiere más de 16 bits en binario.'


mybin(4321)
````

2. Cree una función llamada `myint` pero que a partir de un número binario con base de 16 bits encuentre el entero correspondiente. Compare su resultado con `int(0b1000011100001)`

````python
# Aqui va su código

def myint(b: str) -> int:

  """
  La función retorna la representación decimal de un número binario.
  """

  if len(b) <= 18:  # 16 caracteres de los bits y dos de '0b'

    if '0b' in b:
      n = 0
      for exp, bit in enumerate(b[:1:-1]):  # b[:1:-1]='1000011100001_b0'
        n += int(bit) * 2 ** exp
    else:
      n = 0
      for exp, bit in enumerate(b[::-1]):  # b[:1:-1]='1000011100001'
        n += int(bit) * 2 ** exp

    return n
  else:
    return f'{b} posee más de 16 bits.'


myint('0b1000011100001')
````

3. Generar una función llamada `number64` para flotantes de precisión doble (64bits) donde a partir de un número binario encuentre el valor real. Compruebe su solución usando

```
number64("0100000000111011100100001111111111111111111111111111111111111111")
```



````python
# Aqui va su código

def number64(b : str) -> float:

  """
  Representación decimal de un número binario de precisión doble.
  """

  if 0 < len(b) <= 64:

    s = int(b[0])
    exp = myint(b[1:12])
    frac_mant = sum([int(j) * 2 ** (- i) for i, j in enumerate(b[12::])])

    return (-1) ** s * ((1 + frac_mant) * 2 ** (exp - 1023))

  else:
    return f'Cantidad de bits excedidas.'


number64("0100000000111011100100001111111111111111111111111111111111111111")
````

4. ***(Opcional)*** Escriba una función que convierta un número con base de 8 bits en hexadecimal, compruebe su respuesta con `hex` e imprimalo también en binario.

Observación: ¿la pregunta puede ser formulada de otra manera? No la termino de entender.

````python
# Aqui va su código

def hexadeci(n: int) -> str:

  """
  Número en base hexadecimal.
  """

  dig = {'10':'A', '11':'B', '12':'C', '13':'D', '14':'E', '15':'F'}

  hex = ''
  while n > 0:

    if (n % 16) >= 10:
      hex += dig[str(n % 16)]
      n //= 16
    else:
      hex += str(n % 16)
      n //= 16

  return '0x' + hex[::-1]


hexadeci(15882)
````
