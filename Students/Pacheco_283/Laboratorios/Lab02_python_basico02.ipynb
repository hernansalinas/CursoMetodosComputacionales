1. Elaborar un algoritmo e implementar en python, el factorial de un número entero `n` ingresado por el usuario.

1. Si el número es un entero deberá retornal el factorial del número.
2. Si el número es negativo deberá aparecer un mensaje tipo string con el mensaje  "el número no puede ser negativo"

3. Si el número es pertenece a los reales(float) el mensaje será, el número no puede ser real.


     ### Ejemplo de Ejecución
    ```python

    >>> factorial(3)
        6

    >>> factorial(1987123)
        =???
```


````python
# Factorial de un número entero.

def factorial(n: int):

  """
  Dado n natural, la función determina
  n! = n * (n-1) * (n-2)*...*[n-(n-k)]*...* 2 * 1
  """

  try:

    if type(n) == int and n < 0:
      raise ValueError("Valor inapropiado.")

    elif type(n) != int:
      return "No se aceptan objetos diferentes a enteros positivos."

    else:

      factorial = 1

      while n > 0:
        factorial *= n
        n -= 1

      return factorial

  except ValueError:
    return "El número no puede ser negativo."



factorial(10)
````

````python
# Divisores de un número.

def divisores(n: int):

  """
  Se almacenan los divisores de n en una lista.
  Por ejemplo, los divisores de n=6 son:
  divisores = [1, 2, 3, 6]
  """

  return [i for i in range(1, n) if n % i == 0]


def suma_divisores(n: int) -> int:

  """
  Docstring!
  """

  sum = 0

  for i in divisores(n):
    sum += i

  return sum

````

````python
# Números defectivos.

def numeros_defectivos(n: int):

  """
  Números defectivo : la suma de los divisores propios es menor que el número.
  """

  return suma_divisores(n) < n

````

````python
# Números abundantes

def numeros_abundantes(n: int):

  """
  Números abundantes: la suma de los divisores propios es mayor que el número.
  """

  return suma_divisores(n) > n

````

````python
# Números semiperfectos

def numeros_semiperfectos(n: int):

  """
  Docstring!
  """

  sum = 0

  for i in divisores(n)[::-1]:
    sum += i

    if sum > n:
      sum -= i

    elif sum == n:
      return True

  return False

````

````python
# Números primos

def numeros_primos(n: int) -> bool:

  """
  Docstring!
  """

  if n == 1:
    return False

  return len(divisores(n)) == 1

````

3. Determine si un número `n` entero ingresado por el usuario es un [palíndromo](https://en.wikipedia.org/wiki/Palindromic_number), (Retorne `True` en caso afirmativo y `False` en caso contrario)

```python


    >>> palindromo(3333333)
        True

    >>> palindromo(2323)
        True
        
    >>> palindromo(1111349111111)
        False
```




````python
# Palíndromo

def palindromo(n: int) -> bool:

  """
  Docstring!
  """

  dig_numero = list(str(n))

  return dig_numero == dig_numero[::-1]

````

[Quicksort](https://es.wikipedia.org/wiki/Quicksort).


```python


<<< v = [22, 32, 42, 12, 22, 31, 41, 11, 12, 232, 24, 12, 22]

<<< w = print(quicksort(v))
```

Sólo para comprobar tu código,  puedes hacer uso del comando sort de python.
```python
<<< b = [22, 32, 42, 12, 22, 31, 41, 11, 12, 232, 24, 12, 22]

<<< print(b.sort())

<<< [11, 12, 12, 12, 22, 22, 22, 24, 31, 32, 41, 42, 232]
```

````python

# Reordenamiento de arreglos

def quicksort(arr):

  """
  Docstring!
  """

  if len(arr) <= 1:
    return arr

  else:
    pivot = arr[len(arr) // 2]

    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quicksort(left) + middle + quicksort(right)

````
