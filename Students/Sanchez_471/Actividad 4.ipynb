```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
url = 'https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv'
df = pd.read_csv(url)
```

```python
df['AvgTempCelcius'] = ((df.AvgTemperature) - 32) / 1.8

Bogota = df[df["City"] == "Bogota"].copy()
Barcelona = df[df["City"] == "Barcelona"].copy()

x = Bogota["Year"]
y = Bogota["AvgTempCelcius"]
X = Barcelona["Year"]
Y = Barcelona["AvgTempCelcius"]
Max_Bogota = y.value_counts().idxmax()
Max_Barcelona = Y.value_counts().idxmax()

plt.plot(x,y, label = ('Bogota, la mayor temperatura fue de: '+str(round(Max_Bogota,2)))+'C', color = 'b')
plt.plot(X,Y, label = ('Barcelona, la mayor temperatura fue de: '+str(round(Max_Barcelona,2)))+'C', color = 'r')
plt.legend()
plt.ylabel("Temperatura en celcius $C$")
plt.xlabel("Ao")
plt.grid()
plt.show()
```
