**PUNTO 1**

```python
def suma(M1,M2):

    if np.shape(M1) == np.shape(M2):
        M = np.zeros((np.shape(M1)))
        M[0:np.shape(M)[0] + 1 , 0:np.shape(M)[1] + 1] = M1[0:np.shape(M)[0] + 1 , 0:np.shape(M)[1] + 1] + M2[0:np.shape(M)[0] + 1 , 0:np.shape(M)[1] + 1]
        return M
    
    else:
        raise Exception("Las dimensiones no coinciden")
   
```

```python
suma(M1,M3), M1 + M3
```

**PUNTO 2**

```python
def multiplicacion(M1,M2):

  if np.shape(M1)[1] == np.shape(M2)[0]:
    M = np.zeros((np.shape(M1)[0],np.shape(M2)[1]))
    M[0:np.shape(M)[0] + 1 , 0:np.shape(M)[1] + 1] = np.dot(M1[0:np.shape(M1)[0] + 1,:],M2[:,0:np.shape(M2)[1] + 1])
    return M

  else:
    raise Exception("Las dimensiones no coinciden")
```

```python
multiplicacion(M1,M2), np.matmul(M1,M2)
```

**PUNTO 3**

```python
def transpuesta(M1):
    M = np.zeros((np.shape(M1)[1],np.shape(M1)[0]))
    I, J = np.arange(0,np.shape(M1)[0]),np.arange(0,np.shape(M1)[1])

    for i in I:
        for j in J:
            M[j][i] = M1[i][j]
    return M
```

```python
transpuesta(M2), M2.T
```

**PUNTO 4**

```python
R = lambda x: np.array(([np.cos(x), -np.sin(x)], [np.sin(x), np.cos(x)]))
Rot = lambda x,J: np.matmul(R(x),J)
```

```python
x,J = np.pi/9 , np.array([1,1])

JJ = Rot(x,J.T).T

J,JJ
```

```python
plt.quiver(0,0,J[0],J[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector original')
plt.quiver(0,0,JJ[0],JJ[1], angles='xy', scale_units='xy', scale=1,  color='r', label='Vector rotado')
plt.legend()
plt.grid()
plt.xlim(0,1.5)
plt.ylim(0,1.5)
plt.show()
```
