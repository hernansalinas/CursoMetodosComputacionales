2. El potencial Lennard-Jones es un modelo simple que aproxima la interacción entre un par de moléculas neutras. Este está definido como:

$$
V(r) = 4\epsilon \left[ \left( \frac{\sigma}{r} \right)^{12} - \left( \frac{\sigma}{r} \right)^{6} \right]
$$

Donde:
- $V(r)$ es el potencial en función de la distancia interatómica $r$.
- $\epsilon$ es la profundidad del pozo de potencial, es decir, la energía mínima que describe la interacción entre dos partículas.
- $\sigma$ es la distancia a la cual el potencial es cero, también conocido como el parámetro de distancia.
- $r$ es la distancia entre las dos partículas.

`````python
#Primer Punto
v_m = int(input('Ingrese una velocidad para el tranceunte: '))
a_p = int(input('Ingrese una aceleracion para el policia: '))
t_igual = 2*(v_m/a_p)
t = np.arange(0,t_igual + t_igual/2)
x_f = v_m*t
x_p = (1/2)*(a_p*(t)**2)
x_f_igual = v_m*t_igual

plt.plot(t,x_f, 'g-', label = 'Persona')
plt.plot(t,x_p, 'b-', label = 'Policia')
plt.plot(t_igual,x_f_igual, 'ro', label = 'Punto de encuentro')
plt.xlabel('Tiempo (t)')
plt.ylabel('Posicion (m)')
plt.title('Persecusion policial')
plt.legend()
plt.grid()
plt.show()
`````

`````python
#Punto 2
sigma_ = 100 #funciona para cualquier valor de sigma
def potencial_LJ(r,sigma = sigma_, epsilon= 1):

  return 4*epsilon*((sigma/r)**12 - (sigma/r)**6)
`````

`````python
r = np.linspace(sigma_,2.5*sigma_,500)
pot = potencial_LJ(r)
r_m = (2**(1/6))*sigma_ #es 2**(1/6) no 1/16


plt.plot(r,pot)
plt.plot(r_m,potencial_LJ(r_m), 'ro', label = 'Punto de equilibrio')
plt.title('Potencial de Lennard Jones')
plt.legend()
plt.xlabel('r')
plt.ylabel('Potencial')
plt.grid()
plt.show()

print('\n',f'El minimo de potencial de Lennard Jones es: {pot.min()}, el cual es aproximadamente {potencial_LJ(r_m)}','\n', f'así entonces el minímo de potencial o punto de equilibrio se da con un r = {r_m}')
`````
