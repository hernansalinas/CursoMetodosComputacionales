```python
import numpy as np
import matplotlib.pyplot as plt

def calcular_presion(profundidad_maxima, n_pasos=100):
    profundidades = np.linspace(0, profundidad_maxima, n_pasos)
    densidad_fluido = 1000   # kg/m^3 (para agua)
    aceleracion_gravitacional = 9.8 # m/s^2
    presion_atmosferica = 101325 # Pa (presin atmosfrica estndar al nivel del mar)
    presiones = presion_atmosferica + densidad_fluido * aceleracion_gravitacional * profundidades
    return profundidades, presiones

profundidad_maxima = 30
profundidades, presiones = calcular_presion(profundidad_maxima)

plt.figure(figsize=(8, 6))
plt.plot(profundidades, presiones, 'g-')
plt.xlabel('Profundidad (m)', fontsize=14)
plt.ylabel('Presin (Pa)', fontsize=14)
plt.title('Variacin de Presin con la Profundidad', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)

presion_maxima = presiones[-1]
presiones_atmosfericas = presion_maxima / 101325
label = f'{presiones_atmosfericas:.2f} atm'

plt.legend([label], fontsize=14)
plt.show()
plt.savefig("activdad_02_graf.png.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import json

g = 9.81
hc = 3.2
# theta = 50
# vo = 10.5
# d = 10
def encesto(theta, vo, d):
    x = np.linspace(0, d, 1000)
    t = x / (vo * np.cos(np.radians(theta)))
    y = 2 + vo * np.sin(np.radians(theta)) * t - 0.5 * g * t**2
    return x, y

x, y = encesto(theta, vo, d)
tf = d / (vo * np.cos(np.radians(theta)))
yf = 2 + vo * np.sin(np.radians(theta)) * tf - 0.5 * g * tf**2

if np.abs(yf-hc)<=0.1 and any(y > hc):
    print("El jugador encest.")
else:
    print("El jugador no encest.")

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Trayectoria')
plt.axhline(hc, color='red', linestyle='--', label='Altura del aro')
plt.xlabel('Distancia (m)')
plt.ylabel('Altura (m)')
plt.title('Movimiento parablico del lanzamiento de la pelota')
plt.legend()
plt.grid(True)
plt.show()
```
