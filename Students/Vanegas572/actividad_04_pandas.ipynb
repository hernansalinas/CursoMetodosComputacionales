```python
import pandas as pd
import matplotlib.pyplot as plt
```

1. Leer el url

```python
url = "https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv"
```

```python
df = pd.read_csv(url)
df
```

2. Agregar una nueva columna con la temperatura en grados centr√≠gados

```python
df["AvgTempCelcius"] = (df.AvgTemperature-32)/1.8
df
```

3. Vamos a realizar una gr√°fica donde se muestre la temperatura en funci√≥n de los a√±os para la ciudad de Bogot√° y para la de Barcelona

```python
#comienzo sacando un dataframe con la informaci√≥n de bogot√°
a = df[df["City"]=="Bogota"]
a
```

```python
#otro con la de barcelona
b = df[df["City"]=="Barcelona"]
b
```

4. Sacar la temperatura m√°xima de ambas ciudades

```python
#comienzo sacando la temperatura m√°xima para bogot√°
mbogota = a["AvgTempCelcius"].max()
mbogota
```

```python
#y saco la fila donde se ubica ese m√°ximo para poner extraer su informaci√≥n
a√±o_maximob = a.loc[a["AvgTempCelcius"] == mbogota]
a√±o_maximob
```

```python
#igual para barcelona
mbarcelona = b["AvgTempCelcius"].max()
mbarcelona
```

```python
a√±o_maximobar = b.loc[b["AvgTempCelcius"] == mbarcelona]
a√±o_maximobar
```

5. Gr√°fica donde se ilustran tambi√©n la temperatura m√°xima de cada ciudad

```python
plt.plot(a["Year"],a["AvgTempCelcius"], color = "rebeccapurple", label="Bogota")
plt.plot(a√±o_maximob["Year"],a["AvgTempCelcius"].max(), marker = "o", color="darkorange")
plt.plot(b["Year"],b["AvgTempCelcius"], color = "darkgreen", label="Barcelona")
plt.plot(a√±o_maximobar["Year"],b["AvgTempCelcius"].max(), marker = "o", color="orangered")
plt.title("Temperatura en funci√≥n del a√±o")
plt.ylabel("Temperatura ¬∞C",labelpad=10)
plt.xlabel("A√±os",labelpad=10)
plt.legend()
plt.grid()
```
