Operaciones matricial con regresión lineal.

Cuando se desea hacer una regresion lineal multivariada, existen diferentes formas de realizar la regresión, una de ellas es encontrar el error cuadrático medio y miniminzar el conjunto de parámetros de la regresión emplando algún metodo de minimización, otra forma es encontrar la expresión analítica en forma matricial para enoncontrar el conjunto de parametros de la regresión. En este último caso, la regresion lineal viene dada por la siguiente expresion generalizada:


- $\Theta = (X^T X)^{-1} X^T y$

donte $\Theta$ es la matriz de parametros asociados a la regresión lineal, X la matriz de características y y los valos esperados, así, si tenemos una conjunto de datos en N dimensiones la recta, plano o hyperplano puede ser calculado como:

$Y=\Theta^T X$


En el caso de dos dimensiones, tenemos lo siguiente, Supongamos que X, y Y presentan un comportamiento lineal, y vienen dados por:


- Sea $\vec{X}=x^1, x^2, x^3, x^4, ..., x^m$
- Sea $\vec{Y}=y^1, y^2, y^3, y^4, ..., y^m$

Para determinar la ecuacion de la recta, hay que conocer los párametros $\theta_0 $, $\theta_1$, que cumplen que :   $y^{i} = \theta_1 x^{i} + \theta_0$ la anterior expresion puede ser escrita como:

\begin{equation}
\begin{bmatrix}
y^{1}\\
y^{2}\\
\vdots\\
y^{m}\\
\end{bmatrix}=
\begin{bmatrix}
\theta_0 & \theta_1 \\
\end{bmatrix}
\begin{bmatrix}
1 & 1 & \cdots & 1 \\
x_1^{1} & x_1^{2} & \cdots & x_1^{m} \\
\end{bmatrix}
\end{equation}


Para determinar la matriz de theta entonces se puede aplicar la ecuación:

- $\Theta = (X^T X)^{-1} X^T y$

donde:
\begin{equation}
X=
\begin{bmatrix}
1 & x_1^{1} \\
1 & x_1^{2} \\
\vdots & \vdots \\
1 & x_1^{m}
\end{bmatrix}
\end{equation}

y


\begin{equation}
y=
\begin{bmatrix}
y^1 \\
y^2 \\
\vdots\\
y^m
\end{bmatrix}
\end{equation}

1) Construir un programa que permita multiplicar dos matrices

````python
A = np.random.randint(1,100,(6,5))
B = np.random.randint(1,100,(5,6))
A,B
````

````python
def producto_matrices(A,B):
  m,nA = np.shape(A)
  nB,p = np.shape(B)
  if nA == nB:
    producto = np.zeros((m,p))
    for k in range(0,m):
      for i in range(0,p):
        producto[k,i]= np.sum(A[k] * B[:,i])
  else:
    return print("Las matrices no se pueden multiplicar")
  return producto
````

Cosntruir un programa que permita construir una regresion lineal simple empleando lo que sigue:


````python
N = 10
x1 = np.linspace(0, 10, N)
noise=2*np.random.random(N)-1
y = 0.9*x1 + noise
````

````python
"""construyo la matriz X: la columna de x1 y de unos"""

# columna de unos
columna_1= np.ones_like(x1)


X=np.zeros((N,2))
X[:,0]=columna_1
X[:,1]= x1

X= np.matrix(X)
X
````

se aplica la ecuacion: - $\Theta = (X^T X)^{-1} X^T y$

````python
theta = (X.T @ X).I @ X.T @ y
````

ahora le ecuacion $Y=\Theta^T X$

````python
Y = theta @ X.T
````

````python
""" Voy a construir la regresion lineal simple"""
plt.scatter(x1,y,label="Puntos obtenidos")
plt.plot(x1,Y.T, color="red",label="Regresion lineal")
plt.xlabel("Valores de X")
plt.ylabel("Valores de Y")
plt.legend()
plt.grid()
plt.show();
````

Construir un programa que permita calcular una regresion bidimensional y hacer su representación .

````python
N = 20
x1 = np.linspace(-1, 1, N)
x2 = np.linspace(-2, 2, N)
noise1=2*np.random.random(N)-1
noise2=2*np.random.random(N)-1
y = 2*(x1+noise1) - 3*(x2+noise2) + 0.0 #+ 4*np.random.random(100)

X1, X2 = np.meshgrid(x1,x2)
Y = 2*X1 - 3*X2 + 0.0

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
#surf = ax.plot_surface(X1,X2,Y)
scatter = ax.scatter(x1, x2, y,"-")
````

Construir un programa que permita realizar eliminacion de Gauss- Jordan de una matriz

------------------------------------------------

````python
def gauss_jordan(B):
  def intercambiar_filas(M,i,j):
    A = M.copy()
    M[j,:] = A[i,:]
    M[i,:] = A[j,:]

    return M

  def escalar_matrix(M,lam,i):
    copia = M.copy()
    copia[i,:] = lam * M[i,:]

    return copia

  def suma_producto(M,lam,i,j):
    copia = M.copy()
    copia[j,:] = lam * M[i,:] +  M[j,:]
    return copia


  def pivote(B):
    # la funcion convierte el valor del primera fila y primera  columna en 1
    for i in range(B.shape[0]):
      if B[i,0]!= 0:
        B=intercambiar_filas(B,0,i)
      else:
        B=intercambiar_filas(B,-1,i)
    if B[0,0]>0:
        B =escalar_matrix(B,1/B[0,0],0)
        return B
    else:
      B =escalar_matrix(B,1/B[0,0],0)
      return B
  B=pivote(B)

  # convertir a ceros los numeros que estan  debajo del primer pivote
  for i in range(1,B.shape[0],1):
      if B[i,0]!=0:
        B=suma_producto(B,-B[i,0],0,i)

  def mat(B):
  # la funcion mat convierte los elementos de la diagonal de una matriz en unos y convierte los numeros debajos de esos unos se convierten en ceros
    B=pivote(B)

  # convertir a ceros los numeros que estan  debajo del primer pivote
    for i in range(1,B.shape[0],1):
      if B[i,0]!=0:
        B=suma_producto(B,-B[i,0],0,i)
    return B

  for i in range(1,B.shape[0]):
    if B[i,i] !=0:
      B[i::1,i::1] = mat(B[i::1,i::1])

   # convierte en ceros los numeros por encima de la diagonal
  for j in range(1,B.shape[1]-1):
    B = suma_producto(B,-B[j-1,j],j,j-1)

  B = suma_producto(B,-B[0,B.shape[1]-2],B.shape[1]-2,0)

  return B
````
