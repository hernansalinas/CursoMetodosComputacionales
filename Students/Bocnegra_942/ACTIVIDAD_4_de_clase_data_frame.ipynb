```python
import pandas as pd
```

```python
url2 ="https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv"
df2=pd.read_csv(url2)


#df2["AvgTempCelcius"]=df2.AvgTemperature.
df2
```

```python
df2.info()
```

```python
df3= df2.copy()
df3
```

```python
df3["AvgTempCelcius"]=(df3["AvgTemperature"]-32)/1.8
df3
```

```python
import matplotlib.pyplot as plt


# Crear un grfico
plt.figure()
plt.scatter(df3['Year'][1::2], df3['AvgTempCelcius'][1::2])
plt.plot(df3['Year'][1::2], df3['AvgTempCelcius'][1::2], marker='o')  # Utiliza 'o' para marcar puntos en la lnea
plt.xlabel('AO')
plt.ylabel('TEMP EN CELCIUS')
plt.title('Ao vs Temperatura Bogot')
plt.grid(True)  # Agregar una cuadrcula al grfico
plt.show()




```

```python
# Crear un grfico
plt.figure()
plt.scatter(df3['Year'][0::2], df3['AvgTempCelcius'][0::2])
plt.plot(df3['Year'][0::2], df3['AvgTempCelcius'][0::2], marker='o')  # Utiliza 'o' para marcar puntos en la lnea
plt.xlabel('AO')
plt.ylabel('TEMP EN CELCIUS')
plt.title('Ao vs Temperatura Barcelona')
plt.grid(True)  # Agregar una cuadrcula al grfico
plt.show()

```
