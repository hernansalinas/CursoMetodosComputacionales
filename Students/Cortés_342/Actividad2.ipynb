$$

- $ V(r) $ es el potencial en función de la distancia interatómica $ r $.

`````python
def LJ (r,a,e):
  V = (4*e) *(((a/r)**12)-((a/r)**6))
  return(V)


A = float(input('Ingrese el parámetro de distancia: '))
E = float(input('Ingrese la profundidad del pozo: '))
r = np.linspace(0.9,2.5*A,1000)
V = LJ(r, A , E)

fig, ax = plt.subplots()
ax.plot(r, V )

ax.set(xlabel='Distancia', ylabel='Potencial',
       title='Distancia-Potencial')
ax.grid()


plt.show()
`````
