```python
#Definimos variables
import matplotlib.pyplot as plt
import random

x = 0
y = 0
i = 0
r =float()

x_vals = list()
y_vals = list()

#Se crea el ciclo para repetir cinco mil veces
while(i < 5000):
	r = random.uniform(0,1)
	#Condicional para definir la transformacin segn el numero aleatorio
	if r <= 0.01:
		x_nuevo = 0
		y_nuevo = 0.16*y
	elif 0.01 < r and r <= 0.86:
		x_nuevo = (0.85*x)+(0.04*y)
		y_nuevo = (-0.04*x)+(0.85*y)+1.6
	elif 0.86 < r and r<= 0.93:
		x_nuevo = (0.2*x)-(0.26*y)
		y_nuevo = (0.23*x)+(0.22*y)+1.6
	elif 0.93 < r and r <= 1:
		x_nuevo = (-0.15*x) + (0.28*y)
		y_nuevo = (0.26*x) + (0.24*y) + 0.44

	#Guardamos los datos nuevos en las listas creadas
	x_vals.append(x_nuevo)
	y_vals.append(y_nuevo)

	#Actializamos los valores de X y Y
	x = x_nuevo
	y = y_nuevo

	#Aumentamos una unidad al contador
	i = i+1

plt.figure(figsize=(6, 10))
plt.scatter(x_vals, y_vals, s=0.1, color='green')
plt.title("Helecho de Barnsley")
plt.axis("off")
plt.show()
```
