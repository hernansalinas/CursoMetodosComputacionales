## Realiza la programacion del siguiente helecho

### 1. Inicialización:
- Empezar con un punto inicial en `(x, y) = (0, 0)`.

### 2. Crear listas vacías:
- Crear una lista `x_vals` para guardar los valores de `x`.
- Crear una lista `y_vals` para guardar los valores de `y`.

### 3. Generar puntos repetidamente:
- Repetir el siguiente proceso una cierta cantidad de veces (por ejemplo, 50,000 veces):
  
  a. **Generar un número aleatorio `r` entre 0 y 1**.
  
  b. **Decidir qué transformación aplicar** usando condicionales basados en `r`:
     - Si `r <= 0.01`, aplicar la **Regla 1**:
       - `x_nuevo = 0`
       - `y_nuevo = 0.16 * y`
       
     - Si `0.01 < r <= 0.86`, aplicar la **Regla 2**:
       - `x_nuevo = 0.85 * x + 0.04 * y`
       - `y_nuevo = -0.04 * x + 0.85 * y + 1.6`
       
     - Si `0.86 < r <= 0.93`, aplicar la **Regla 3**:
       - `x_nuevo = 0.2 * x - 0.26 * y`
       - `y_nuevo = 0.23 * x + 0.22 * y + 1.6`
       
     - Si `0.93 < r <= 1`, aplicar la **Regla 4**:
       - `x_nuevo = -0.15 * x + 0.28 * y`
       - `y_nuevo = 0.26 * x + 0.24 * y + 0.44`
       
  c. **Guardar los nuevos valores `x_nuevo` y `y_nuevo`**:
     - Añadir `x_nuevo` a la lista `x_vals`.
     - Añadir `y_nuevo` a la lista `y_vals`.

  d. **Actualizar `x` y `y`** con los valores nuevos:
     - `x = x_nuevo`
     - `y = y_nuevo`

### 4. Dibujar los puntos:
- Usar una librería gráfica para dibujar los puntos almacenados en las listas `x_vals` y `y_vals`.


```python
# Ahora dibujamos el helecho usando los puntos que hemos generado
```

````python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random
````

````python
n = 100000

x = 0.
y = 0.
x_nuevo = 0.
y_nuevo = 0.
x_vals = []
y_vals = []
colors = np.random.rand(n)
for i in range(n):
  r = np.random.rand()
  if r <= 0.01:
    x_nuevo = 0.
    y_nuevo = 0.16 * y
  if r > 0.01 and r <= 0.86 :
    x_nuevo = 0.85 * x + 0.04 * y
    y_nuevo = -0.04 * x + 0.85 * y + 1.6
  if r > 0.86 and r <= 0.93 :
    x_nuevo = 0.2 * x - 0.26 * y
    y_nuevo = 0.23 * x + 0.22 * y + 1.6
  if r > 0.93 and r <= 1 :
    x_nuevo = -0.15 * x + 0.28 * y
    y_nuevo = 0.26 * x + 0.24 * y + 0.44
  x = x_nuevo
  y = y_nuevo
  x_vals.append(x_nuevo)
  y_vals.append(y_nuevo)

#x_vals = np.array(x_vals) <---- Esto no permite la correcta visualización.
#y_vals = np.array(x_vals)
#print(x_vals)
#print(y_vals)
plt.style.use('dark_background')
plt.figure(figsize=(6, 10))
plt.scatter(x_vals, y_vals, s=0.1, c=colors)
plt.title("Helecho de Barnsley")
plt.axis("off")
plt.show()
````
