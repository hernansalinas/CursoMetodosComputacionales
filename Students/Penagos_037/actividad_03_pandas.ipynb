```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv")

df.insert(6, "AvgTempCelcius", (df["AvgTemperature"]-32)/1.8)

df
```

```python
dfBogota=df[df["City"]=="Bogota"]
dfBarcelona=df[df["City"]=="Barcelona"]

plt.plot(dfBogota["Year"], dfBogota["AvgTempCelcius"], label="Bogot")
plt.plot(dfBarcelona["Year"], dfBarcelona["AvgTempCelcius"], label="Barcelona")
plt.legend()
plt.ylabel("temperatura ($C^\circ$)")
plt.xlabel("ao")
plt.show()
```

```python
maxTempBogot=dfBogota.max()["AvgTempCelcius"]
maxTempBarcelona=dfBarcelona.max()["AvgTempCelcius"]

print(f"""La temperatura mxima de Bogot fu de {maxTempBogot}
La temperatura mxima de Barcelona fu de {maxTempBarcelona}""")
```
