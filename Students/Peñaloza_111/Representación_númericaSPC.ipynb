1. Escriba una función que se llame `mybin` que reciba un número entero real de 16 bits y realice su conversión en código binario, compare su resultado con `bin(4321)`

````python
def mybin(num):
  num_keep = num
  binary_str = ""
  for i in range(16):
    binary_str = str(num % 2) + binary_str
    num //= 2
  if int(binary_str) == int(str(bin(num_keep))[2:]):
    print("algoritmo correcto")
  return binary_str
mybin(4321)
````

2. Cree una función llamada `myint` pero que a partir de un número binario con base de 16 bits encuentre el entero correspondiente. Compare su resultado con `int(0b1000011100001)`

````python
def myint(binary_str):
    decimal_value = 0
    for i, bit in enumerate(reversed(binary_str)):
        if bit == '1':
            decimal_value += 2**i
    return decimal_value

# Ejemplo de uso:
result = myint(str(1000011100001))
print(f"El número binario {1000011100001} es {result} en decimal.")
````

3. Generar una función llamada `number64` para flotantes de precisión doble (64bits) donde a partir de un número binario encuentre el valor real. Compruebe su solución usando

```
number64("0100000000111011100100001111111111111111111111111111111111111111")
```



````python
def number64(binary_str):
    if len(binary_str) != 64:
        raise ValueError("La cadena binaria debe tener exactamente 64 bits")

    # Extraer los elementos de un binario
    signo = int(binary_str[0])
    exponente_str = binary_str[1:12]
    mantisa_str = binary_str[12:]

    # Convertir a enteros
    exponente = int(exponente_str, 2) - 1023
    mantisa = int(mantisa_str, 2) / 2**52  # Normalizar la mantisa

    # Calcular el valor
    valor = (-1)**signo * (1 + mantisa) * 2**exponente
    print(f"{valor}")
    return valor

# Ejemplo de uso
binary_num = "0100000000111011100100001111111111111111111111111111111111111111"
number64(binary_num)


````
