```python
import random
import matplotlib.pyplot as plt



x = [0]
y = [0]

def f1(x,y):
  xn = 0
  yn = 0.16*y[-1]
  x.append(xn)
  y.append(yn)

def f2(x,y):
  xn = 0.85*x[-1] + 0.04*y[-1]
  yn = -0.0044*x[-1] + 0.85*y[-1] + 1.6
  x.append(xn)
  y.append(yn)


def f3(x,y):
  xn = 0.2*x[-1] - 0.26*y[-1]
  yn = 0.23*x[-1] + 0.22*y[-1] + 1.6
  x.append(xn)
  y.append(yn)

def f4(x,y):
  xn = -0.15*x[-1] + 0.28*y[-1]
  yn = 0.26*x[-1] + 0.24*y[-1] + 0.44
  x.append(xn)
  y.append(yn)



def generator():
    opciones = [f1, f2, f3, f4]
    probabilidad = [1, 85, 7, 7]
    for i in range(0,100000):
      [f] = random.choices(opciones, weights=probabilidad)
      f(x,y)
generator()
plt.plot(x,y,".", ms=0.1)
```
