### Factorial
1. Elaborar un algoritmo e implementar en python, el factorial de un número entero `n` ingresado por el usuario.

1. Si el número es un entero deberá retornal el factorial del número.
2. Si el número es negativo deberá aparecer un mensaje tipo string con el mensaje  "el número no puede ser negativo"

3. Si el número es pertenece a los reales(float) el mensaje será, el número no puede ser real.


     ### Ejemplo de Ejecución
    ```python

    >>> factorial(3)
        6

    >>> factorial(1987123)
        =???
```


````python
def factorial(n):
  total = 1
  if n < 0:
    print("no pueden ser numeros negativos ")
  elif (n % 2) != 0:
    print("no pueden ser numeros reales ")
  else:
    n = int(n)
    for i in range(1,n+1):
      total = i * total
    return total

factorial(float(input("escriba el n factorial ")))

````

````python
divisores_a = []
divisores_b = []
def numeros_amigos(a,b):
  for i in range(1,b):
    if b % i == 0:
      divisores_b.append(i)
  for i in range(1,a):
    if a % i == 0:
       divisores_a.append(i)
  test1 = sum(divisores_a)
  test2 = sum(divisores_b)
  if test1 == b and test2 ==a:
    print(f"{a} y {b} son numeros amigos")
  else:
    print(f"{a} y {b} no son numeros amigos")

  return divisores_b, divisores_a

numeros_amigos(int(input("Escriba un numero a")), int(input("Escriba un numero b")))
````

3. Determine si un número `n` entero ingresado por el usuario es un [palíndromo](https://en.wikipedia.org/wiki/Palindromic_number), (Retorne `True` en caso afirmativo y `False` en caso contrario)

```python


    >>> palindromo(3333333)
        True

    >>> palindromo(2323)
        True
        
    >>> palindromo(1111349111111)
        False
```




````python
def palindromo(n):
  n_tw = int((str(n))[::-1])
  if n / n_tw == 1:
    print(f"el numero {n} es un palindromo")
  else:
    print(f"el numero {n} no es un palindromo")

  return n_tw
palindromo(int(input(f"Ingrese un numero")))
````

[Quicksort](https://es.wikipedia.org/wiki/Quicksort).

Sólo para comprobar tu código,  puedes hacer uso del comando sort de python.
```python
<<< b = [22, 32, 42, 12, 22, 31, 41, 11, 12, 232, 24, 12, 22]

<<< print(b.sort())

<<< [11, 12, 12, 12, 22, 22, 22, 24, 31, 32, 41, 42, 232]
```

````python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        less = [x for x in arr[1:] if x <= pivot]
        greater = [x for x in arr[1:] if x > pivot]
        return quicksort(less) + [pivot] + quicksort(greater)


numbers = [22, 32, 42, 12, 22, 31, 41, 11, 12, 232, 24, 12, 22]
sorted_numbers = quicksort(numbers)
print(sorted_numbers)
````
