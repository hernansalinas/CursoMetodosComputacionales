21.0  Cálculo de $\pi$. Empleando la siguiente seria :

\begin{equation}
\frac{\pi}{4} = 1 -\frac{1}{3}+\frac{1}{5}-\frac{1}{7}+\frac{1}{9}-...
\end{equation}

Construya una función llamada pi_number en el que entrado el valor de k,(número de términos de la serie ) calcule el valor de pi. Si k es negativo deberá retornar un error como una variable tipo string.

Ejemplo de ejecución:
```python
pi_number(10)
```

````python

#Definimos la función pi_number.
def pi_number(k):
  #Creamos la variable convergencia, que llevara la cuenta del valor de la sumatoria en cada iteración.
  convergencia = 0
  if k<0:
    #si k es negativvo entonces devuelve un error
    error = "numero negativo invalido"
    print(f"{error}")
  else:
    #Se crea un ciclo For que recorre desde 0 hasta la cantidad de terminos, este es el indice de la sumatoria.
      for i in range(0,k):
        #Creamos una variable que se llama "serie", que genera terminos de la serie de Leibniz.
        serie = ((-1)**i)/((2*i)+1)
        #Sumamos a la convergencia el valor de la serie par obtener un valor aproximado de π/4
        convergencia = convergencia + serie
      return convergencia

pi_number(int(input("Ingrese el numero de terminos de la serie  ")))


````
