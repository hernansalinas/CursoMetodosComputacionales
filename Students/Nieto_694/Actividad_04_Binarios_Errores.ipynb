### Activity 1.0: 

Implement a function that get the binary representation of an integer with name mybin, the algorithm should return 
a string with 8 characters, the maximun number that represent the funcition is integer of 8 bit. The name of function is 
mybin

Ejexample : 

b=mybin(x)

b='00101000'

````python
def mybin(x):
    binario = ''
    while x // 2 != 0:
        binario = str(x % 2) + binario
        x= x // 2
    return str(x) + binario
````

Para cada dígito de la representación binaria, se multiplica por 2 elevado a la posición del dígito (empezando desde cero en la posición más a la derecha).

El resultado de la multiplicación se suma al número decimal equivalente.

Una vez se han multiplicado y sumado todos los dígitos binarios, el resultado final es el número decimal equivalente.

````python
def number(BIN):
    BIN=str(BIN)
    decimal=0
    posicion =0
    for i in BIN[::-1]:
      BIN=int(BIN)
      if i == "1":
        decimal += 2**posicion
      posicion +=1
    decimal=float(decimal)
    return decimal
````
