# **1. Movimiento rectilíneo uniforme**

1. Una persona conduce una moto con rapidez constante $v_m$, mientras que un policia lo hace con aceleración constante $a_p$ (iniciando desde el reposo). Suponga que ambos movimientos inician en $t=0$, haga una gráfica para el movimiento de la moto y del policia, donde el usuario ingrese la velocidad (moto) y aceleración (policia). Además determine el instante $t$ en el cual se encuentran ambos. Ambos movimientos iniciarán en $x_i = 0$

**Pruebe**: para $v_m$ = 15m/s y $a_p$ = 3 m/s², el tiempo deberia ser 10 s


$$x_f = x_i + v_m\cdot t$$
$$x_f = x_i + v_o\cdot t + \frac{1}{2}a_p\cdot t^2$$
---

`````python
def encuentro(vm, vo, ap):
  t = (2*(vm-vo))/ap
  return t
`````

`````python
def pos_encuentro(v,te):
  x_e = v*te
  return x_e
`````

`````python

v1 = float(input("Ingresa la velocidad del cuerpo 1 (MRU): "))
v2 = float(input("Ingresa la velocidad inicial del perseguidor: "))
a = float(input("Ingresa la aceleración del perseguidor: "))
#x_perseguidor = float(input("Ingresa la posición inicial del perseguidor: "))
#x_fugitivo = float(input("Ingresa la posición inicial del fugitivo: "))
`````

`````python
t = encuentro(v1, v2, a)
t
`````

`````python
x = pos_encuentro(v1,t)
x
`````

`````python
t1 = np.linspace(0,t+t/10,500)
#print(t1)
xf = v1*t1
`````

`````python
xp = (1/2)*a*(t1**2)
`````

`````python
plt.plot(t1,xp, 'b-', label='Policía')
plt.plot(t1,xf, 'r-', label='Fugitivo')
plt.legend()
plt.title("Movimiento de la moto y policía vs tiempo.")
plt.grid()
#plt.xlim([-1,t+t/10])
#plt.ylim([-3,xp[len(500)]])
plt.ylabel("Posición del policía y el fugitivo (m)")
plt.xlabel("Tiempo (s)")
plt.plot(t,x, marker="o", color = 'm')
plt.show()
`````

# **2. Potencial de Lennard-Jones**

2. El potencial Lennard-Jones es un modelo simple que aproxima la interacción entre un par de moléculas neutras. Este está definido como:

$$
V(r) = 4\epsilon \left[ \left( \frac{\sigma}{r} \right)^{12} - \left( \frac{\sigma}{r} \right)^{6} \right]
$$

Donde:
- $V(r)$ es el potencial en función de la distancia interatómica $r$.
- $\epsilon$ es la profundidad del pozo de potencial, es decir, la energía mínima que describe la interacción entre dos partículas.
- $\sigma$ es la distancia a la cual el potencial es cero, también conocido como el parámetro de distancia.
- $r$ es la distancia entre las dos partículas.

a. Escriba una función para encontrar modelar el potencial LJ.


b. Usando $\sigma = \epsilon = 1$ demuestre que la distancia de equilibrio $r_m$ es aproximadamente $r_m = 2^{1/16}\sigma$. Para esto realice un gráfico del potencial, especificando los ejes. Trunque la serie en $r_c = 2.5\sigma$.


c. Verifique que $r_m = 2^{1/16}\sigma$ se cumple para diferentes valores de $\sigma$, dejando $\epsilon$ constante.

`````python
#print("Ingresa en las siguientes celdas los valores de:")
e = 1
s = 1
`````

`````python
def leonard_jones(epsilon, sigma, r):
  a = (sigma/r)**(12)
  b = (sigma/r)**(6)
  V = 4*epsilon*((a)-(b))
  return V
`````

`````python
distancia_part = np.linspace(0.9,2.5*s,500)
#distancia_part
`````

`````python
V_prima = leonard_jones(e,s,distancia_part)
#V_prima
`````

`````python
rm = 2**(1/16)
rm
`````

`````python
plt.plot(distancia_part,V_prima, 'b-', label='Potencial de Leonard-Jones')
plt.legend()
plt.title("Potencial de Leonard-Jones vs distancia entre partículas.")
plt.grid()
#plt.xlim([-1,t+t/10])
#plt.ylim([-3,xp[len(500)]])
plt.ylabel("Potencial de Leonard-Jones ")
plt.xlabel("Distancia entre las partículas (sigmas)")
plt.axvline(rm*s, color = 'red')
plt.axvline((2**(1/6))*s, color = 'green')
plt.show()
`````
