1. Escriba una función que se llame `mybin` que reciba un número entero real de 16 bits y realice su conversión en código binario, compare su resultado con `bin(4321)`

````python
# Aqui va su código

import math

def mybin(x):
  binario=[]
  while x >= 2:
    if x%2 == 0:
      binario.append(0)
    if x%2 != 0:
      binario.append(1)
    x = math.trunc(x/2)
  binario.append(1)
  binario_new = binario[::-1]
  resultado_str = ''.join(str(bit) for bit in binario_new)
  resultado_int = int(resultado_str)
  return resultado_int
````

````python
x = int(input(print("Ingrese un número entero real de 16 bits en base decimal")))

if -2**15 <= x <= 2**15 - 1:
  resultado1 = mybin(x)
  print("El número",x," equivale al número",resultado1,"en base binaria")
  print(bin(x))
else:
  print("El número ingresado no es de 16 bits")
````

2. Cree una función llamada `myint` pero que a partir de un número binario con base de 16 bits encuentre el entero correspondiente. Compare su resultado con `int(0b1000011100001)`

````python
# Aqui va su código

def myint(x):
  digitos = [int(a) for a in str(x)]
  l = len(digitos)
  numeros = []
  k = 0
  for i in digitos:
    d = i*2**(l-k-1)
    numeros.append(d)
    k = k+1
  decimal = 0
  for t in numeros:
    decimal = decimal + t
  return decimal
````

````python
x = int(input(print("Ingrese un número de 16 bits en base binaria")))
n = str(x)
if len(n)<=16:
  resultado2 = myint(x)
  print("El número",x,"en binario equivale al número",resultado2,"en base decimal")
else:
  print("El número ingresado no es de 16 bits")
````

3. Generar una función llamada `number64` para flotantes de precisión doble (64bits) donde a partir de un número binario encuentre el valor real. Compruebe su solución usando

```
number64("0100000000111011100100001111111111111111111111111111111111111111")
```



````python
# Aqui va su código

def number64(n):

  if len(n) != 64:
    raise ValueError("El número debe tener 64 bits")

  signo = int(n[0])
  exponente = n[1:12]
  mantisa_str = str(n[12:])

  exp = int(exponente,2)-1023

  mantisa = 1
  for i, bit in enumerate(mantisa_str):
    if bit == '1':
      mantisa += 2**-(i+1)

  valor_real = (-1)**signo * 2**exp * mantisa

  return valor_real
````

````python
n = str(input(print("Ingresa el binario de 64 bits")))
resultado3 = number64(n)
print(resultado3)
````
