### Factorial
1. Elaborar un algoritmo e implementar en python, el factorial de un número entero `n` ingresado por el usuario.

1. Si el número es un entero deberá retornal el factorial del número.
2. Si el número es negativo deberá aparecer un mensaje tipo string con el mensaje  "el número no puede ser negativo"

3. Si el número es pertenece a los reales(float) el mensaje será, el número no puede ser real.


     ### Ejemplo de Ejecución
    ```python

    >>> factorial(3)
        6

    >>> factorial(1987123)
        =???
```


````python
def factorial(n):
  if n < 0:
    print("El número no puede ser negativo")
  elif isinstance(n,float) == True:
    print("El número no puede ser un real no entero")
  else:
    f=1
    for t in range(1,n+1):
      f=f*t
  return f

n=5
result = factorial(n)
print(result)
````

````python
#Función para los divisores de n#

def divisores(n):
  divisores = []
  for i in range(1,n):
    if n%i == 0:
      divisores.append(i)
  return divisores
````

````python
n=int(input(print("Ingresa un número para calcular sus divisores")))
result = divisores(n)
print(result)
````

````python
#2 - Números defectivos#

def numeros_defectivos(d):
  sum = 0
  l = divisores(d)
  for t in l:
    sum = sum + t
  if sum < d:
    return True
  else:
    return False

d=int(input(print("Ingresa un número para saber si es defectivo")))
result2 = numeros_defectivos(d)
print(result2)
````

````python
#3 - Números abundantes#

def numeros_abundantes(a):
  sum = 0
  l = divisores(a)
  for t in l:
    sum = sum + t
  if sum > a:
    return True
  else:
    return False

a=int(input(print("Ingresa un número para saber si es abundante")))
result3 = numeros_abundantes(a)
print(result3)
````

````python
#4 - Números semiperfectos#

from itertools import combinations

def numeros_semiperfectos(s):
  l = divisores(s)
  for r in range(1, len(l)+1):
    for comb in combinations(l,r):
      if sum(comb) == s:
        return True
  return False

s = int(input(print("Ingresa un número para saber si es semiperfecto")))
result4 = numeros_semiperfectos(s)
print(result4)
````

````python
#5 - Números perfectos#

def numeros_perfectos(p):
  l = divisores(p)
  sum = 0
  for t in l:
    sum = sum + t
  if sum == p:
    return True
  else:
    return False

p=int(input(print("Ingresa un número para comprobar si es perfecto")))
result5 = numeros_perfectos(p)
print(result5)
````

P2. Retornar los 10 primeros pares de numeros perfectos, semiperfectos, amigos



````python
#7#

def numeros_amigos(a,b):
  da = divisores(a)
  db = divisores(b)
  sa = 0
  sb = 0
  for i in da:
    sa = sa + i
  for j in db:
    sb = sb + j
  cond1 = 0
  cond2 = 0
  if a == sb:
    cond1 = 1
  if b == sa:
    cond2 = 1
  if cond1 + cond2 == 2:
    return True
  else:
    return False

a = int(input(print("Ingresa el primer número amigo")))
b = int(input(print("Ingresa el segundo número amigo")))
respuesta = numeros_amigos(a,b)
print(respuesta)
````

3. Determine si un número `n` entero ingresado por el usuario es un [palíndromo](https://en.wikipedia.org/wiki/Palindromic_number), (Retorne `True` en caso afirmativo y `False` en caso contrario)

```python


    >>> palindromo(3333333)
        True

    >>> palindromo(2323)
        True
        
    >>> palindromo(1111349111111)
        False
```




````python
import math

def es_palindromo(x):
  numeros = [int(a) for a in str(x)]
  l = len(numeros)
  l_par = int(l/2)
  l_impar = int(math.trunc(l/2))
  ladoi = []
  ladod = []
  if l%2 == 0:
    for i in range(0,l_par):
      ladoi.append(numeros[i])
    for i in range(l_par,l):
      ladod.append(numeros[i])
    ladod2 = ladod[::-1]
    if ladoi == ladod2:
      return True
    else:
      return False
  if l%2 != 0:
    for i in range(0,l_impar):
      ladoi.append(numeros[i])
    for i in range(l_impar+1,l):
      ladod.append(numeros[i])
    ladod2 = ladod[::-1]
    if ladoi == ladod2:
      return True
    else:
      return False
````

````python
x = str(input(print("Ingresa un número para saber si es palíndromo")))
respuesta = es_palindromo(x)
print(respuesta)
````

8 Construir un programa en el que  entrado un arreglo de números se  ordenen de forma ascendente, Ver algoritmo
[Quicksort](https://es.wikipedia.org/wiki/Quicksort).


```python


<<< v = [22, 32, 42, 12, 22, 31, 41, 11, 12, 232, 24, 12, 22]
<<< def quicksort(v):

<<<     return v
<<< w = print(quicksort(v))
<<< [11, 12, 12, 12, 22, 22, 22, 24, 31, 32, 41, 42, 232]

```

Sólo para comprobar tu código,  puedes hacer uso del comando sort de python.
```python
<<< b = [22, 32, 42, 12, 22, 31, 41, 11, 12, 232, 24, 12, 22]

<<< print(b.sort())

<<< [11, 12, 12, 12, 22, 22, 22, 24, 31, 32, 41, 42, 232]
```

````python
def quicksort(n):
  if len(n) <= 1:
    return n
  else:
    pivote = n[0]
    menores = [x for x in n[1:] if x <= pivote]
    mayores = [x for x in n[1:] if x > pivote]
    return quicksort(menores) + [pivote] + quicksort(mayores)

n=[3,6,1,8,4,7,3,80,12]
n_ordenado = quicksort(n)
print(n_ordenado)
````
