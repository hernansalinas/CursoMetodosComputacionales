### Factorial
1. Elaborar un algoritmo e implementar en python, el factorial de un número entero `n` ingresado por el usuario.

1. Si el número es un entero deberá retornal el factorial del número.
2. Si el número es negativo deberá aparecer un mensaje tipo string con el mensaje  "el número no puede ser negativo"

3. Si el número es pertenece a los reales(float) el mensaje será, el número no puede ser real.


     ### Ejemplo de Ejecución
    ```python

    >>> factorial(3)
        6

    >>> factorial(1987123)
        =???
```


````python
def factorial(n):
    if n<0:
        return 'el número no puede ser negativo'
    elif type(n)==float:
        return 'No puede ser float'
    elif n==0:
        return 1
    else:
        lista = []
        contador= 1
        fact= 1
        while contador<=n:
            lista.append(contador)
            contador+=1
        for num in lista:
            fact*=num
        return fact
````

````python
factorial(2)
````

````python
# Numero defectivo
def numeros_defectivos(a):
    l = list(range(1,a+1))
    divisores= []
    i=0
    for num in l:
        if a%num==0 and a!=num:
            divisores.append(num)
    for div in divisores:
        i+=div
    if i<a:
        return True
    else:
        return False
````

````python
# Numero semiperfecto
from itertools import combinations

def numeros_semiperfectos(a):

    l= list(range(1,a+1))
    divisores = []
    resultado =  {}
    for num in l:
        if a%num==0 and a!=num:
            divisores.append(num)

    for i in range(1,len(divisores)+1):
        for comb in combinations(divisores,i):
            if sum(comb)==a:
                return True

    return False
````

    La rutina se debera llamar numeros_amigos




P2. Retornar los 10 primeros pares de numeros perfectos, semiperfectos, amigos



````python
# Numeros amigos
def numeros_amigos(a,b):

    l= list(range(1,a+1))
    div_a = []
    i=0
    for num in l:
        if a%num==0 and a!=num:
            div_a.append(num)
    for num in div_a:
        i+=num

    l2= list(range(1,b+1))
    div_b= []
    suma= 0
    for num_ in l2:
        if b%num_==0 and b!=num_:
            div_b.append(num_)
    for num in div_b:
        suma+=num

    if a==suma and b==i:
        return True
    else:
        return False
````
