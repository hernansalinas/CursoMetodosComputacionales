# ***Interpolacion Lineal***

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
#x= np.array([1,2,3,4])
#y= np.array([1,3,4,6])

x = np . array ([1,2,3,4,5,6])
y = np . array ([2,4,6,7,9,13])
n = 0
plt.figure(figsize=(12,5))

while n < len(x)-1:
  xo = x[n]
  yo = y[n]
  m = (y[n+1]-y[n])/(x[n+1]-x[n])
  x_ = np.linspace(x[n],x[n+1],4)
  y_ = yo + (m*(x_- xo))
  n +=1

  plt.plot(x,y, "o")
  plt.xlabel("x")
  plt.ylabel("f(x)")
  plt . title (' Interpolación lineal')
  plt.grid()
  plt.plot(x_,y_)
```

```python
x = np . array ([1,2,3,4,5,6])
y = np . array ([2,4,6,7,9,13])

x_ = np.linspace (1 ,6 ,100)
y_ = interpolate.interp1d (x ,y )

plt.figure(figsize=(12,5))
plt.plot (x ,y ,'o', x_ , y_ ( x_ ) ,'-')
plt.title (' Interpolación lineal')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid ()
```
