```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Lectura de el dataset desde la url, con el comando pd.read_csv.

```python
# Lectura del arichivo csv atraves del link
url = "https://raw.githubusercontent.com/hernansalinas/autogradesMetodosComputacionales/main/Dataset/Temperature_Bogota_Barcelona.csv"
df = pd.read_csv(url)
df
```

Adicion de una nueva columna AvgTempCelcius con la temperatura en grados celcius.

$ C=(F-32)/1.8$

```python
df["AvgTempCelcius"] = (df.AvgTemperature - 32)/1.8
df
```

Realizacion un grfico de AvgTempCelcius como funcin del ao para la ciudad de Bogot.

```python
# Filtro del DataFrame solo para la ciudad de Bogota
tf = df.City == "Bogota"
df1 = df[tf].copy()

# Graficacion
plt.figure(figsize=(12,5))
plt.plot(df1.Year , df1.AvgTempCelcius , "red")
plt.title("TEMPERATURA VS TIEMPO")
plt.ylabel("Temperatura en Bogota (C)")
plt.xlabel("Aos")
plt.ylim(8,15)
plt.show()

```

# Realizacion un grfico de AvgTempCelcius como funcin del ao para la ciudad de Barcelona.

```python
# Filtro del DataFrame solo para la ciudad de Barcelona
tf = df.City == "Barcelona"
df2 = df[tf].copy()

# Graficacion
plt.figure(figsize=(12,5))
plt.plot(df2.Year , df2.AvgTempCelcius , "red")
plt.title("TEMPERATURA VS TIEMPO")
plt.ylabel("Temperatura en Barcelona (C)")
plt.xlabel("Aos")
plt.ylim(13,18)
plt.show()
```

# Determinacion la mayor temperatura para Bogot y Barcelona.

```python
m_bo= df1["AvgTempCelcius"].max()
m_ba= df2["AvgTempCelcius"].max()
print(f"La mxima temperatura registrada en Bogot fue de {m_bo:.2f}C")
print(f"La mxima temperatura registrada en Barcelona fue de {m_ba:.2f}C")
```
