```python
import scipy as sc
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x,a: a*np.sin(x) + (np.tan(x))**2
#x = np.arange(5,180,5)
x_ = np.linspace(0,1,1000)
plt.axvline(x=(np.pi/2), color="k", ls="-.")
plt.plot(x_,f(x_,1))
plt.grid()

h=lambda x: sp.sin(x)+(sp.tan(x))**2
x=sp.Symbol("x")
h(x)

```

```python
dx=x_[1]-x_[0]
a = 1
g = sc.misc.derivative(f, x_,dx=dx,args=(a,))

```

```python
plt.plot(x_,g)
plt.axvline(x=(np.pi/2), color="k", ls="-.")
plt.grid()
```

```python
f = lambda x: x**2
fp = lambda x, h: (1/(2*h))*(f(x+h) - f(x-h))

x = np.linspace(-100, 100, 1000)
h = x[1] - x[0]

# Vectorizamos la funci√≥n fp
vectorized_fp = np.vectorize(fp)

plt.plot(x, vectorized_fp(x, h), label="f prima")
plt.plot(x, f(x), color="red", label="$x^2$")
plt.plot(x,2*x, label="2x")
plt.legend()
plt.grid()

plt.show()





```

como se puede ver, la grafica derivada obtenida por el metodo de los 3 puntos y la derivada obtenida manualmente es la misma, por ende el metodo de los 3 puntos hallo la derivada correctamente

```python
f= lambda x0: a*np.sin(x0)+np.tan(x0)**2
fp=lambda x0,h: (1/(2*h))*(f(x0+h)-f(x0-h))

x0=np.linspace(-0.4,1,100)
h=x0[1]-x0[0]



plt.plot(x0, fp(x0, h), label="f prima")
plt.plot(x0, f(x0), color="red", label="$a*np.sin(x)+np.tan(x)**2$")
plt.legend()
plt.grid()

plt.show()
```
