```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
x=np.array([2,5,6])
f=np.array([4,1,2])
```

```python
# Coordenadas

plt.plot(x,f,"o",c="b")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Puntos a interpolar")
plt.grid()
```

```python
x_s= sp.Symbol("x") # x simbolicamente
x_s
```

```python
def Lagrange(j,x):
  l_n=1

  # Params = Polinomio de Lagrange j-esimo
  # n = NÃºmero de puntos
  # x = Puntos a interpolar

  for i in range(0,(len(x))):
    if (i!=j):
     l_n*=(x_s-x[i])/(x[j]-x[i])
  return l_n
```

```python
l_n= Lagrange(0,x)
l_n
```

```python
# Polinomio de Lagrange

def P(x_s,x,f):
  poly=0
  for i in range (0,len(f)):
    poly+= f[i]*Lagrange(i,x)
  return poly
```

```python
P(x_s,x,f)
```

```python
sp.simplify(P(x_s,x,f))
```
