```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# Condiciones iniciales
a=45
thetha = math.radians(a) #en radianes
v0 = 10 # m/s
g = 9.8 # m/s**2
y0 = 0
x0 = 0
```

```python
# Posición en x

def x_m (x,t,v0,thetha):
  cos = np.cos(thetha)
  x = x0+(v0*cos)*t
  return(x)

# Posición en y

def y_m (y,t,v0,thetha):
  sen = np.sin(thetha)
  y = y0+(v0*sen)*t-((g*(t)**2)/2)
  return(y)

tv= 2*v0*np.sin(thetha)/g
t = np.linspace(0,tv,300) # continuidad
x = x_m(0,t,v0,thetha)
y = y_m(0,t,v0,thetha)
```

```python
# Gráfica

plt.plot(x,y,color="red")
#plt.plot(x_max,y_max,"ro")
plt.title("Trayectoria parabólica")
plt.xlabel("x", fontsize=12)
plt.xlim(0,11)
plt.ylim(0,3)
plt.ylabel("y", fontsize=12)
plt.grid(0.2)
plt.show()

```
