````python
import random
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
````

````python
x , y = 0 , 0 #Inicializar con el punto (x,y) = (0,0)
x_values, y_values = [], [] #Genera listas vacías

for i in range (1, 50000): #Genera puntos repetidamente
  r = round(random.uniform(0,1),2)
  if r<= 0.01: #regla 1
    x_nuevo, y_nuevo = 0, 0.16*y
  if 0.01 < r <= 0.86: #regla 2
    x_nuevo, y_nuevo = 0.85 * x + 0.04 * y, -0.04 * x + 0.85 * y + 1.6
  if 0.86 < r <= 0.93: #regla 3
    x_nuevo, y_nuevo = 0.2 * x - 0.26 * y, 0.23 * x + 0.22 * y + 1.6
  if 0.93 < r <= 1: #regla 4
    x_nuevo, y_nuevo = -0.15* x + 0.28* y, 0.26 * x + 0.24 * y + 0.44
  x_values.append(x_nuevo) #Guarda los nuevos valores en las  listas
  y_values.append(y_nuevo)

  x , y = x_nuevo , y_nuevo #Actualiza los valores nuevos
````

````python
plt.figure(figsize=(6, 10))
plt.scatter(x_values, y_values, s=0.1, color = 'green')
plt.title("Helecho de Barnsley")
plt.axis("off")
plt.show()
````
