### El potencial Lennard-Jones es un modelo simple que aproxima la interacción entre un par de moléculas neutras. Este está definido como:

$$
V(r) = 4\epsilon \left[ \left( \frac{\sigma}{r} \right)^{12} - \left( \frac{\sigma}{r} \right)^{6} \right]
$$

Donde:
- $ V(r) $ es el potencial en función de la distancia interatómica $ r $.
- $ \epsilon $ es la profundidad del pozo de potencial, es decir, la energía mínima que describe la interacción entre dos partículas.
- $ \sigma $ es la distancia a la cual el potencial es cero, también conocido como el parámetro de distancia.
- $ r $ es la distancia entre las dos partículas.

a. Escriba una función para encontrar modelar el potencial LJ.


b. Usando $\sigma = \epsilon = 1$ demuestre que la distancia de equilibrio $r_m$ es aproximadamente $r_m = 2^{2/16}\sigma$. Para esto realice un gráfico del potencial, especificando los ejes. Trunque la serie en $r_c = 2.5\sigma$.


c. Verifique que $r_m = 2^{2/16}\sigma$ se cumple para diferentes valores de $\sigma$, dejando $\epsilon$ constante.


`````python

#a
def potencial_LJ (radio, epsilon, sigma):
  a = (sigma/radio)**12
  b = (sigma/radio)**6
  return 4*epsilon*(a-b)

#b
sig = 1
R = np.linspace(1,2.5*sig-0.5,10000)


V = potencial_LJ(R,1,1)
min = np.argmin(V)
print(V)
plt.plot(R,V)
plt.axvline((2**(2/16)))
# Esta linea es la que aparece al lado izquierdo de la grafica, y vemos que no coincide con el pozo potencial.
# al derivar la expresion anterior con resecto a la distancia, encontre que el pozo potencial vendria dado por r = (2*(sigma**2))**(1/6)
# ahora veamos que:
plt.axvline(2**(1/6),linewidth=4, color='b')
#
plt.title("Potencial vs radio")
plt.xlabel("Radio")
plt.ylabel("V(r)")



`````

`````python
#probemoslo con otro valor de sigma para comprobar
# sigma = 1.1

sigm = 1.1
R = np.linspace(0.99999,2.5*sigm,10000)

V = potencial_LJ(R,1,sigm)
min = np.argmin(V)
print(V)
plt.plot(R,V)
plt.axvline((2**(2/16)*sigm))
plt.axvline((2**(1/6))*sigm,linewidth=4, color='b')
#
plt.title("Potencial(1) vs radio")
plt.xlabel("Radio")
plt.ylabel("V(r)")
`````

a. Limpie los datos, eliminando columnas en *(in)* y cambie todos los nombres al español.

`````python
import pandas as pd

# Dataset 1
# URL of the webpage
url = "https://worldpopulationreview.com/country-rankings/average-height-by-country"

# Read the HTML content
html_content = pd.read_html(url)

alturas = html_content[2]
alturas
`````

`````python
#a) comenzaremos eliminando las columnas en (in)
alturas1 = alturas.drop(columns=['Boys 19 (in)','Girls 19 (in)'])

# Ahora cambiemos las columnas a español
alturas1.columns = ["Pais", "Altura promedio, niños, 19 años (cm)", "Niñas 19 (cm)", "Niños 15 (cm)", "Niñas 15 (cm)", "Niños 10 (cm)", "Niñas 10 (cm)", "Niños 5 (cm)", "Niñas 5 (cm)"]

alturas_organizadas = alturas1.sort_values(by = "Niños 15 (cm)",ascending=False)

alturas_organizadas
`````

b. Realice un gráfico de barras, donde se muestre una comparación de la estatura media para niños y niñas en ambas categorias. Esto para los primeros 10 paises con mayor y menor altura, además agregue a Colombia.

`````python
#b)
colombia = alturas_organizadas[alturas_organizadas['Pais']=='Colombia']

primeras_10 = alturas_organizadas.head(11)
ultimas_10 = alturas_organizadas.tail(11)

alturas_e = pd.concat([primeras_10, colombia, ultimas_10])

alturas_e
`````
