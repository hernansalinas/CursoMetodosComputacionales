1. densidad con respecto a la altura

```python
import numpy as np
import matplotlib.pyplot as plt

ro0 = 1.29
lamda = 7482.2
def ro(h):
  return ro0*np.exp(-h/lamda)

h = np.linspace(0,30000)

plt.title("Funcion densidad con respecto a la altura")
plt.plot(h,ro(h),label = "Densidad")
plt.xlabel("Altura(m)")
plt.ylabel("Densidad(kg/m^3)")
plt.legend(loc=1)
plt.grid()
```

2. Diseñar un programa que resuelva la ecuación diferencial que describe el movimiento del paracaidista, para los siguientes parámetros:

$m=60kg$, $A=0.3m^2$, $h=30km$.

$m=60kg$, $A=0.3m^2$, $h=40km$.

$m=60kg$, $A=0.3m^2$, $h=50km$.

$m=60kg$, $A=0.3m^2$, $h=60km$.

$m=60kg$, $A=0.3m^2$, $h=70km$.

PARA
$m=60kg$, $A=0.3m^2$, $h=30km$.


```python
from scipy.integrate import  odeint
```

```python
def func(x,t,params):
  m = params[0]
  g = params[1]
  k = params[2]
  lamda = params[3]

  dxdt = x[1]

  dvdt = -9.8 + (k*(x[1]**2)/m)*np.exp(-x[0]/lamda)

  return [dxdt,dvdt]
```

```python
x_0 = 30000
x_1 = 0
x0 = np.array([x_0,x_1]) #posicion y velocidad


t = np.linspace(0,200,10000)
```

```python
m = 60
g = 9.8
A = 0.3
k = 1.29* A *0.6/(2)
lamda = 7482.2

params = [m,g,k,lamda]
```

```python
x = odeint(func,x0,t, args=(params,))
```

```python
x1 = x[:,0]
x2 = x[:,1]
```

```python
plt.figure()
plt.plot(x1,-x2)
plt.xlabel("x")
plt.ylabel("v(t)")
plt.xlim(0,30000)
plt.grid()
plt.title("Velocidad con respecto a la posicion")
```

```python
plt.figure()
plt.plot(t,x1)
plt.xlabel("t")
plt.ylabel("x(t)")
plt.xlim(0,200)
plt.ylim(0,30000)
plt.grid()
plt.title("posicion con respecto al tiempo")
```

```python
plt.figure()
plt.plot(t,-x2)
plt.xlabel("t")
plt.ylabel("v(t)")
plt.xlim(0,250)
plt.grid()
plt.title("velocidad con respecto al tiempo")
```

```python
vmax = max(-x[:,1])
print("velocidad maxima = ", vmax)
```
