```python
import numpy as np 
import matplotlib as plt
from pylab import *
import math as m
```

CAMPO PRODUCIDO POR UNA CARGA PUNTUAL

```python
q1=1                            # Carga 
r1=np.array([0.0001,0.0001])    # Posicion de la carga en R2
r2=np.array([1,1])    # Punto en el que se quiere hayar el campo en R2
ke=1                            # Constante


# Función para hallar el campo en un punto del espacio
def Campo_E (q1,r1,r2):     
  global ke
  r=r2-r1
  ur=r/np.linalg.norm(r)  # Vector unitario del campo 
  
  E=((ke*q1)/(np.linalg.norm(r))**2)*ur
  return (E)
```

```python
campo=Campo_E (q1,r1,r2)  
```

```python
# Gráfico de prueba para un punto donde 
# se quiere hallar el campo producio por la carga

plot(r1[0],r1[1],"ro" )
plot(r2[0],r2[1],"o")
quiver(r2[0],r2[1],campo[0],campo[1])
xlim(0,3)
ylim(0,3)
xlabel("Eje X")
ylabel("Eje Y")
```

```python

# Arreglo de vectores para hallar el campo en distintos puntos del espacio

x=np.arange(-2,3,0.5)
y=np.arange(-2,3,0.5)
puntos=np.zeros((len(x)*len(y),2))
 
c=0
for i in x:
  for j in y:
    puntos[c]=np.array([i,j])
    c=c+1
```

```python
# Arreglo que contiene el valor del campo en los puntos del arreglo anterior
campos=np.zeros((len(x)*len(y),2))
k=0
for i in puntos:
   campos[k]=Campo_E(q1,r1,i)
   k=k+1

```

```python
# Grafica del campo producido por una carga puntual en el espacio

quiver(puntos[:,0],puntos[:,1],campos[:,0],campos[:,1],scale=10)
plot(r1[0],r1[1],"o")


xlabel("Eje X")
ylabel("Eje Y")
```

DIPOLO

```python
r1=np.array([-3,0.00001]) # Posicion de la carga 1
r2=np.array([3,0.00001])  # Posicion de la carga 2

q1=2                # Valor de la carga 1
q2=-2               # Valor de la carga 2

# Arreglo de vectores para hallar el campo en distintos puntos del espacio
x=np.arange(-5,6,0.5)
y=np.arange(-5,6,0.5)
puntos=np.zeros((len(x)*len(y),2))
c=0
for i in x:
  for j in y:
    puntos[c]=np.array([i,j])
    c=c+1

# Valor del campo producido por la carga 1 en los puntos del arreglo anterior
campos_q1=np.zeros((len(x)*len(y),2))
k=0
for i in puntos:
   campos_q1[k]=Campo_E(q1,r1,i)
   k=k+1


# Valor del campo producido por la carga 2 en los puntos del arreglo anterior
campos_q2=np.zeros((len(x)*len(y),2))
k=0
for i in puntos:
   campos_q2[k]=Campo_E(q2,r2,i)
   k=k+1


# El campo total en un punto es la suma del campo producido por la carga 1 mas
# el campo producido por la carga 2

campo_dipolo=np.zeros((len(x)*len(y),2))

for i in range(0,len(campo_dipolo)):
  campo_dipolo[i]=campos_q1[i]+campos_q2[i]


quiver(puntos[:,0],puntos[:,1],campo_dipolo[:,0],campo_dipolo[:,1],scale=10)
plot(r1[0],r1[1],"o")
plot(r2[0],r2[1],"o")

xlabel("Eje X")
ylabel("Eje Y")

```
