$\frac{a_c}{a} = 1 + \frac{b}{a}\epsilon_b -\frac{c}{a}\epsilon_c $ 

$\frac{a_c}{a} = 1 + \frac{b}{a} \max(\epsilon_b,\epsilon_c) $ 

Si $b\approx c$, el valor de a es un valor muy grande.

## Error de una serie
Sea $x$ el valor verdadero y $x^*$ el valor aproximado

- Error absoluto : se define como 
\begin{equation*} 
\epsilon_{abs}= |x-x^*|
\end{equation*}
- Error relativo : es dado por 
\begin{equation*} 
\epsilon_{rel}= \frac{|x-x^*|}{|x|}
\end{equation*}
- Error en series : El error para truncar una serie se toma como
\begin{equation*} 
\epsilon_{aprox}= \left|\frac{nth\hbox{-term}}{\hbox{suma}}\right|< \hbox{eps}
\end{equation*}
La tolerancia normalmente se toma como un número pequeño, por ejemplo `eps` $=10^{-10}$. Note que no se trunca la serie usando $|{nth}\hbox{-term}|<$ eps,   usar esta forma puede conducir a errores debido a que no se compara con el valor de la suma (un millon compardo a uno es grande, pero comparado a diez mil millones es pequeño).

# Realizar  el calculo de la funcion sin(x) empleando la serie: 

\begin{equation}
\sin(x) = \sum_{i=1} ^ N (-1)^{n+1} \frac{x^{2n-1}} {(2n-1)!}
\end{equation}


````
#n-esimo termino de la serie de Taylor del sin de x
import numpy as np
import math as mt
import matplotlib.pylab as plt
def _sin(x, n):
  num = (-1) ** (n + 1) * x ** (2 * n - 1)  #Calculo del numerador
  den = mt.factorial(2 * n - 1)             #calculo del denominador
  
  return num/den             

#ESte rutina realiza  el calculo de la funcion sin, para el funcionamiento 
#adecuado de la rutina es necesario ingresar el valor x a calcular y el numero 
#de terminos de la serie de Taylor que se desean 
def ite_sin(x, N):
  
  suma = 0                          # Inicializacion de la suma
   
  for i in range(1, N ):            # Suma de los N terminos 
  
    suma = _sin(x, i) + suma
  
  return suma
````

## Cancelación sustractiva



**serieSn**. Un ejemplo de la cancelacion substractiva es en la suma de series alternantes. Considere la serie $\mathrm{S}_N$,  escrita  matematicamente de tres formas diferentes:

- 
\begin{equation}
\mathrm{S}_N^{(1)} = \sum_{n = 1}^{2N} (-1)^n \frac{n}{n+1} 
\end{equation}

- Sumando números pares e impares.

\begin{equation}
\mathrm{S}_N^{(2)} = -\sum_{n = 1}^{N}\frac{2n-1}{2n} + \sum_{n=1}^{N} \frac{2n}{2n+1}
\end{equation}

- Combinando las series de una forma analítica

\begin{equation}
\mathrm{S}_N^{(3)} = \sum_{n = 1}^{N}\frac{1}{2n(2n+1)}
\end{equation}



Las tres sumas $\mathrm{S}_N^{(1)}$, $\mathrm{S}_N^{(2)}$ y $\mathrm{S}_N^{(3)}$ son matemáticamente iguales, pero los resultados  pueden tener resultados numéricos diferentes.


- Escribir un programa que realice el cálculo de $\mathrm{S}_N^{(1)}$, $\mathrm{S}_N^{(2)}$ y $\mathrm{S}_N^{(3)}$

````
#This notebook allows to analyze an alternating series, its convergence, graph and relative error with respect to the exact value

#Libraries
import matplotlib.pyplot as plt
import numpy as np

def sn_1(N):

  """

  This function represents an alternating series.

  """

  result = 0
  sign = -1

  #summation
  for i in range(1, 2 * N + 1):
    term =  i / (i + 1)
    result += sign * term
    sign = -sign

  return result

def sn_2(N):

  """

  This function represents, once again, the series sn_1, but instead of 
  alternating, it sums odd and even values separately.

  """

  sum1 = 0
  sum2 = 0

  #summation
  for i in range(1, N + 1):
    par = 2 * i
    sum1 += (par - 1) / par
    sum2 += par / (par + 1)

  return sum2 - sum1

def sn_3(N):

  """

  This function represents the combination of the series in an 
  analytic form.

  """

  result = 0

  #summation
  for i in range(1, N + 1):
    par = 2 * i
    result += 1 / (par * (par + 1))

  return result

def convergencia(x):

  """

  This function returns the difference between a certain value and the previous 
  value to see the number of iterations necessary for that difference to be 
  lesser than 1E-06

  """

  epsilon = 1E-07
  i = 3
  n1 = x(1)
  n2 = x(2)

  #analysis of the current and previous value
  while 3 <= i < 5000:
    if n2 - n1 <= epsilon:
      break
    else:
      n1 = n2
      n2 = x(i)
    i += 1

  return i

def error(x, n):

  """

  This function returns the relative error of an inputted function by comparing
  it to the function sn_3.

  """

  sn3 = sn_3(n)

  return abs((x(n) - sn3) / sn3)

def plot(x, err=False):

  """

  This function creates the values necessary to plot the error of the inputted
  function.
  x is the function that is to be plotted.
  err variable changes whether we want the to plot the error or the function 
  itself. True for normal function, False for the error.
  

  """
  if err:
    A = np.arange(1, 1583, 1)
    m = np.zeros(len(A))

    #change the values of the matrix of zeroes into the errors of matrix A
    for i in range(0, len(A)):
      m[i] = error(x, A[i])

  else:
    A = np.arange(1, 200, 1)
    m = np.zeros(len(A))

    #change the values of the matrix of zeroes into the function, it runs the... 
    #...values of matrix A
    for i in range(0, len(A)):
      m[i] = x(A[i])

  return A, m
````

## **Errores de redondeo**. Caos numérico en un mapa logístico y errores de punto flotante (Errores de redondeo)

> Un ejemplo clásico de caos,  es el comportamiento no lineal en las interaciones de un mapa logistico 

\begin{equation}
x_{n+1}=f(x_n)=rx_n(1-x_n)
\end{equation}

> con $x\in (0,1)$ y $r\in(0,4)$ se pueden producir varios comportamientos sorprendentes.


> ### Problema: 
Encontrar tres formas diferentes de expresar $f(x)$  y calcular la evolución de la misma condición inicial después de cientos de iteraciones. Para este problema, será extremadamente útil  ver sus resultados gráficamente; construya listas de números y llame la libreria matplotlib. 


>#### Hint: valores numéricos de r y x0 que puede tomar:


```python
    >>> r = [1.9, 2.9, 3.1, 3.5, 3.9]

    >>> x0 = 0.6 # any number in [0,1] will do here

    >>> numpoints = 100
       ```
       

````
#This code allows the analysis of the numerical chaos function in 3 different shapes to use in a logistic map.

def caos1(x, r):

  """
  This function represents the equation rx(1-x).
  """

  return r * x * (1 - x)

def caos2(x, r):

  """

  This function represents the equation rx-r(x^2).

  """

  return  r * x - r * x ** 2

def caos3(x, r):

  """

  This function represents the equation r(x^2)(1/(x-1)).

  """

  return r * x ** 2 * (1 / x - 1)

def plot_(f, xo, r):

  """

  This function returns the values for the Y axis for plotting the 
  functions caos1, caos2 and caos3.

  """

  Y = np.zeros((len(r), len(r)))   #Array of zeros with shape (10,10) for y-axis values.
  Y[0] = xo
  for i in range(1, len(r)):  #Iteration to assign y-values according to the function.
    Y[i] = f(Y[i-1], r)
  return Y
````

````
r = np.arange(0.1, 4, 0.4)  #Array with len=10 for x-axis values.
xo = 0.6   #Initial value of the population.
plt.figure()  #Creating a new figure.
plt.plot(r, plot_(caos1, xo, r), 'o', label= 'caos1')  #Plot r and caos1 using color circle markers.
plt.plot(r, plot_(caos2, xo, r), '*', label= 'caos2')  #Plot r and caos2  using color star-shaped markers.
plt.plot(r, plot_(caos3, xo, r), ':', label= 'caos3')  #Plot r and caos3 using : linestyle.
plt.show()
````
