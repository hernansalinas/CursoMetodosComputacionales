```python
"""IMPORTAMOS EL CSV DE DRIVE"""

vc=pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vQAmvlb5eMmziu0Aab3XGEG9aud6Kmy8luQKzU9kkuT65szIsnvozJQI1VSwjEAbpitvCWxa5ly0T5b/pub?output=csv")
```

```python
""" Datos solo para colombia """

vc_mask=vc['country']=="Colombia"   # Mascara para filtrar los datos de la columna country que contengan unicamente Colombia

filtered_vc = vc[vc_mask]           # Se aplica la mascara al dataframe original y se almacena en un nuevo dataframe 

filtered_vc                         #Imprimimos el nuevo dataframe con los datos para Colombia
```

```python
""" Se almacenan en una lista los paises de latinoamerica """

latinoamerica=["Argentina" 
,"Bolivia"
,"Brazil"
,"Chile"
,"Colombia"
,"Costa Rica"
,"Cuba"
,"Ecuador"
,"El Salvador"
,"Guayana Francesa"
,"Granada"
,"Guatemala"
,"Guyana"
,"Haiti"
,"Honduras"
,"Jamaica"
,"Mexico"
,"Nicaragua"
,"Paraguay"
,"Panama"
,"Peru"
,"Puerto Rico"
,"Dominican Republic"
,"Suriname"
,"Uruguay"
,"Venezuela"] 
```

```python

""" Rutina para extraer del dataframe original unicamente los paises de latinoamerica"""


total=[]  # Lista donde se almacenaran los dataframe de cada pais de latinoamerica filtrado del dataframe original

for i in latinoamerica:

  mask=latino_vc.loc[:, "country"]==i    # Mascara para filtrar unicamente los datos de la columna country que coinciden con i (paises de latinoamerica) 
  pais=latino_vc.loc[mask]      # Mascara aplicada para obtener los datos para cada pais de latinoamerica por separado
  total.append(pais)  # Esta lista contiene 26 entradas, cada entrada es un dataframe de los datos para los paises de latinoamerica




""" Como cada pais tiene varios datos segun la fecha en la que se hizo el reporte de vacunados, para obtener el numero total de vacunados
en la ultima fecha reportada para cada país, vasta con buscar en el dataframe de cada pais en la columna total_vaccinations, el elemento maximo, 
ese corresponderá al numero de vacunados a la fecha mas reciente """

vacunados=[] # Creamos una lista para almacenar los total vacunados para cada pais

for i in range(0,len(total)): 
  
  a=total[i]       # Se crea un dataframe para cada pais 
  v=a["total_vaccinations"].max()  # se busca el valor máximo de la columna total_vaccinations para cada país
  vacunados.append(v)    # Se añade en la lista

""" Como los elementos de la lista vacunados esta en el mismo orden de la lista latinoamerica, se crea un dataframe con estas dos listas"""

total_vacunados=pd.DataFrame()  # Dataframe con los páises y su respectivo numero de vacunados a la fecha

total_vacunados["Pais"]=latinoamerica # Columna para los paises
total_vacunados["Vacunados"]=vacunados # Columna para los vacunados a la fecha

""" Se eliminan las entradas NaN, estas corresponden a paises que no tienen reporte
 en el dataframe original, luego se organizan numéricamente los datos"""
total_vacunados=total_vacunados.fillna(0).sort_values("Vacunados") 

```

```python
""" Grafica de comparacion de los paises segun su numero de vacunados, se observa que Brazil tiene el mayor numero de vacunados a la fecha"""
ax = sns.barplot(y = "Pais", x="Vacunados", data = total_vacunados)
```

```python
""" Para que conocer cuales paises tienen entre 1.5 y 2.0 millones de vacunados se aplica la mascara mask"""

mask=((vc.total_vaccinations >=1.5E06) & (vc.total_vaccinations <=2.0E06)) # Mascara aplicada a la columna total_vaccinations del dataframe original con todos los paises
vc_intervalo=vc[mask]  # Se almacenan los paises que cumplen la mascara anterior en la serie vc_intervalo
vc_intervalo
```

```python
""" Como el procedimiento de la mascara anterior se repite el nombre de los paises ya que para cada pais segun la fecha hay distintos 
valores de vacunados totales, vasta con eliminar los paises duplicados y obtenemos los paises que han vacunado entre 1.5 y 2 millones de personas"""

intervalo=vc_intervalo.country.drop_duplicates()
intervalo




```
