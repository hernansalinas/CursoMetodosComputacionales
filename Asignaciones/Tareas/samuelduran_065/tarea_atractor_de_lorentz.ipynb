````python
import scipy.integrate as integrate
import numpy as np
import matplotlib.pyplot as plt
````

````python
#=10
#=8/3
#=10
def atractor(,,,U0):
  def du_dt(U, t):
    return [ *(U[1]-U[0]),
            U[0]*(-U[2])-U[1],
            U[0]*U[1]-*U[2]]
  ts=np.linspace(0,100,10000)
  Us=integrate.odeint(du_dt,U0,ts)
  x=Us[:,0]
  y=Us[:,1]
  z=Us[:,2]
  fig = plt.figure()
  ax = fig.gca(projection="3d")
  ax.plot(x,y,z)
  plt.draw()
  plt.show()
````
