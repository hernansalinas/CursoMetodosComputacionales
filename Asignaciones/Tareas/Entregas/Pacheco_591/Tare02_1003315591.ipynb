```python
import pylab as plt
import numpy as np

def e(x,n):
  g=0
  for i in range(n):
    g+=((-1)**i)*((x**i)/(np.math.factorial(i)))
    print(((-1)**i)*((x**i)/(np.math.factorial(i))))
  return g
```

```python
e(7,1000), np.exp(-7)
```

```python
def e1(x):
  g=0
  a=1
  i=-1
  while a==1:
    i=i+1
    g+=((-1)**i)*((x**i)/(np.math.factorial(i)))
    h=((-1)**(i+1))*((x**(i+1))/(np.math.factorial(i+1)))
    if abs(h)<=abs(g*(10**-17)): a=0
  return g
```

```python
x=np.arange(0,1,0.1)
y=[]
for i in x: 
  y.append(e1(i))
z=np.exp(-x)
plt.plot(x,abs(y-z))
```

```python
import numpy as np
import matplotlib.pylab as plt

from IPython.display import clear_output
from time import sleep
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python

x = np.linspace(0, 2*np.pi, 100) 
y = np.sin(x)
plt.figure(figsize=(15, 12))
plt.plot(x, y, "b.-", markersize = 2.0, linewidth = 2.0, label="f(x) = sin(x)")
plt.plot(x, y/2, "r.-", markersize = 2.0, linewidth = 2.0, label="f(x) = sin(x)")
plt.xlabel("x", fontsize=20)
plt.ylabel("f(x)", fontsize=20)
plt.grid()
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
plt.show()
```

Se define la funcin de  tres maneras: 

```python
def s1(r,x0,n=100):
    s=x0
    y=[]
    for i in range(n):
       s=r*s*(1-s)
       y.append(s)
    return s,y
def s2 (r,x0,n=100):
    s=x0
    y=[]
    for i in range(n):
        s=r*s-s**2
        y.append(s)
    return s,y
def s3(r,x0,n=100):
    s=x0
    y=[]
    for i in range(n):
        s=s*(r-s)
        y.append(s)
    return s,y
```

```python
x=[1.9,2.9,3.1,3.5,3.9]
plt.figure(figsize=(20,10))
for i in x:
    s=str(i)
    plt.plot(np.arange(0,100),s1(i,0.9,100)[1],"o",label=("r="+s))
plt.legend()
plt.grid()
plt.title("Iteraciones dependiendo de r para x0=0.9")
plt.show()
```

```python
x=[1.9,2.9,3.1,3.5,3.9]
plt.figure(figsize=(20,10))
for i in x:
    s=str(i)
    plt.plot(np.arange(0,100),s2(i,0.9,100)[1],"o",label=("r="+s))
plt.legend()
plt.grid()
plt.title("Iteraciones dependiendo de r para x0=0.9")
plt.show()
```

```python
x=[1.9,2.9,3.1,3.5,3.9]
plt.figure(figsize=(20,10))
for i in x:
    s=str(i)
    plt.plot(np.arange(0,100),s3(i,0.9,100)[1],"o",label=("r="+s))
plt.legend()
plt.grid()
plt.title("Iteraciones dependiendo de r para x0=0.9")
plt.show()
```

Las iteraciones se estabilizan para los valores de r= [1.9,2.9,3.5] y tiene de forma catica en r=3.9
Se pueden apreciar las pequeas variaciones de una grfica a otra por las respectivas modificaciones de las funciones. Los clculos variarian dependiendo de estos cambios.
